## <span style="color: blue;">**Texas Special Education: Landscape & Open Data Directory**</span>

## **Claude 제작**

Here's the interactive dashboard for Texas Special Education open-source data. It's organized into 8 tabs:

> 🔗 **Texas Special Education: Landscape & Open Data Directory** (https://claude.ai/public/artifacts/80c59c9f-c46f-4737-a5a3-febcbd6867b4)

![Texas_Special_Ed](https://raw.githubusercontent.com/pilsunchoi/images4/main/26-1.png)

- **Overview** — Key metrics at a glance + enrollment trend & disability distribution previews
- **Enrollment** — 10-year student count and rate changes from 2015-16 through 2024-25 (dual Y-axis chart)
- **Disability** — All 13 IDEA disability categories shown via horizontal bar chart + donut chart
- **By Region** — Sortable table of all 20 ESC regions with click-to-highlight and top-10 bar chart
- **Placement** — Least Restrictive Environment (LRE) breakdown by educational setting
- **Outcomes** — Graduation/dropout trend lines + SPP Indicator achievement status (Met/Not Met)
- **Demographics** — Race/ethnicity comparison (SPED vs. all students) + age group distribution
- **Data Catalog** — 12 open data source cards categorized by State, Federal, Nonprofit, and Research, each with a direct external link to the original dataset

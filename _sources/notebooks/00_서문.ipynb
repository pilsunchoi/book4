# **서문** 

이제는 어떤 과제든, 학교 숙제든 직장 업무든, 일단 AI한테 한 번 물어보고 시작하는 세상이 됐다. 그만큼 요즘의 AI는 훌륭한 답을 실수 없이 뚝딱 내놓는다. 어찌 보면 그 끝판왕이 데이터 분석 분야일 것이다. 불과 1~2년 전까지만 해도 분석의 아이디어나 코딩, 통계학 지식을 묻고 답하는 데 그쳤지만, 지금은 데이터만 넣어주면 AI가 스스로 분석해서 전문가 수준의 보고서나 인터랙티브 대시보드까지 만들어준다. 데이터의 종류도 제한이 없다. 스프레드시트 형식의 정형 데이터는 물론이고, 텍스트, 이미지, 동영상 등 데이터 분석의 지평이 예전과는 비교할 수 없을 정도로 넓어졌다. 이 책은 현 시점 AI의 데이터 분석 능력을 예시적으로 보여주는 데 목적이 있다. 이를 통해 독자들이 AI 데이터 분석의 현재 수준을 가늠하고, 자신의 과제에 활용할 수 있는 실마리를 얻기 바란다.

우리는 생성형 AI에게 다음과 같은 데이터 분석 과제를 예시적으로 시켜봤다.

* 수강생들의 시험성적 분석
* 대통령 지지율 연령별/지역별 변화 분석
* 2025 KBO 정규시즌 팀기록 분석
* 구매력평가 이론 그래프 분석 및 업데이트
* COVID-19 전후 주요국 경제성장률 수집 및 분석
* 호가창/캔들차트 활용 주가상승 확률 추정 시스템 구축
* 의대 쏠림현상 신문기사 댓글 분석
* 서울 아파트 일별 매물건수 변화 감지 분석
* 항공사 승객 만족도 설문조사 분석
* 보스턴 지역 주택가격 회귀분석
* 종부세와 월세의 실증 관계 보도내용 평가
* 대형마트 쇠락 원인 실증 분석 방법론
* 신생아 특례대출과 출산율의 실증 관계 보도내용 평가
* 은행 텔레마케팅 데이터 로지스틱 회귀분석
* 트리 모델 기반 타이타닉 탑승객 생존 분류
* 딥러닝 모델 기반 손글씨 분류
* 임금 결정 모형 패널 데이터 분석
* 2018 월드컵 결승 경기 데이터 분석
* 이란 폭격 충격 전파 패턴 추정
* 세계 화장품 교역 트렌드 분석
* 텍사스주 특수교육 현황 및 오픈 데이터 조사

우리는 Claude, ChatGPT, Gemini에게 이상의 주제와 관련해 다양한 형태의 자료를 입력한 다음, 스스로 분석하고 설명해달라고 요청했다. 통계학 지식이나 통계 프로그램 코드를 물어본 게 아니라, AI가 “직접” 분석하도록 했다는 게 이 책의 핵심이다.

우리가 AI에게 제공(업로드)한 자료는 Excel, CSV, JSON 파일 등 정형 데이터도 있지만, 텍스트, 이미지, 동영상 등 비정형인 경우도 많다. 비정형 데이터 분석에서 멀티모달(multimodal) AI 모델의 강력함이 발휘된다.

어떤 경우에는 분석할 데이터를 주지 않고, AI 스스로 데이터를 찾아서 분석해달라고 요구하기도 했다. COVID-19 전후 주요국 경제성장률 수집 및 분석, 세계 화장품 교역 트렌드 분석, 텍사스주 특수교육 현황 및 오픈 데이터 조사 등이 여기에 해당한다.

우리의 분석 요청(프롬프트)도 다양하다. 필요한 분석 방향을 아주 구체적으로 제시한 경우도 있지만, 단순히 “분석해줘”라고 요청하기도 했다. 또한 이미 남이 해놓은 실증분석에 대해 어떻게 생각하는지 물어보기도 했다. 종부세와 월세의 관계, 신생아 특례대출과 출산율의 관계 등 미디어 보도에 대한 평가가 여기에 해당한다.

이처럼 다양한 상황에서 AI가 발휘하는 놀라운 분석 능력을 이 책의 26개 장에서 확인할 수 있다. 막연한 요청만으로도 전문가 수준의 보고서를 1~2분 만에 뚝딱 내놓는다. 분석 및 조사 결과를 인터랙티브 대시보드를 사용해 직관적으로 보여주는 것도 매우 인상적이다.

AI는 뛰어난 능력을 갖췄지만, 동시에 실수를 하기도 하고, 때로는 잘못된 결과를 제시한다는 것도 목격했다. 사실이 아님에도 그럴듯하게 설명하는 이른바 환각(hallucination)에 주의해야 한다. 더욱이 AI는 사용자의 질문 방식이나 맥락을 반영하여 답을 생성한다는 점도 고려해야 한다. 결국 중요한 것은 AI를 효과적으로 활용하는 능력과 함께, 그 결과를 비판적으로 점검하고 선별할 수 있는 연구자의 역량이다.

각 장의 마지막에는 분석에 사용된 파이썬 코드를 수록했다. AI가 분석 결과와 함께 코드를 제공하는 경우도 있지만, 그렇지 않을 때에는 별도로 요청하면 된다. AI의 파이썬 실행 환경(샌드박스)은 보안과 안정성을 위해 `pandas`, `numpy`, `matplotlib`, `scikit-learn` 등 사전에 설치된 라이브러리만 사용할 수 있도록 제한돼 있다. 따라서 사전 설치되지 않은 패키지(예: 이 책 20장의 `linearmodels`)의 경우에는 추가 설치를 하지 못해 우리에게 코드만 주는 상황이 발생할 수 있다. 이런 경우 사용자가 로컬 환경에서 실행하여 그 결과를 AI에게 다시 입력해주는 방식을 취해야 한다.

이 책의 분석에는 (집필 시점 기준) Claude Opus 4.6, GPT-5.2, Gemini 3 Pro를 활용했다. 이들이 이 책의 공동 작업자들이다.

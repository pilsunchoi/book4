### 0. 먼저: 지금 그림이 왜 ‘증거’가 아닌가

연도별로 “상위 분위에서 출산율이 올라갔다”는 건 **패턴(상관)** 이다. 하지만 2024-2025에 신생아 특례대출과 동시에 일어난 다른 변화(금리, 경기, 혼인/이혼, 주택가격, 다른 정책 등)가 상위 분위에 더 강하게 작용했을 수도 있다. 그러면 출산율 상승을 대출정책에 귀속시키면 오판이다.

그래서 필요한 게 “정책이 없었으면(**반사실, counterfactual**) 어떻게 됐을까”를 만드는 구조이고, **DID(Difference-in-Differences)** 는 그걸 만드는 가장 표준적인 방법 중 하나다.

---

### 1. 소득분위 × 연도 DID(이중차분) — “전후 변화의 차이”를 계층별로

#### 1.1 기본 DID의 뼈대

가장 단순한 DID는 이렇게 생긴다.

* 집단: 처리집단(treated) vs 비교집단(control)
* 시점: 정책 전(pre) vs 정책 후(post)
* 핵심:

$$
DID = (Y_{T,post}-Y_{T,pre}) - (Y_{C,post}-Y_{C,pre})
$$

여기서 $Y$는 합계출산율(TFR) 같은 결과변수.

이게 말하는 건:  
“처리집단이 정책 이후 더 늘었는데, 그건 단순한 시간 추세가 아니라 처리집단에서만 추가로 생긴 변화”라는 해석이다.

---

#### 1.2 “소득분위 × 연도”를 DID로 한다는 건 무슨 뜻?

소득 10분위를 그대로 쓰면 ‘집단’이 10개다. 즉 “한 집단 vs 한 집단”이 아니라 **다집단 DID(패널/반복횡단면)** 형태가 된다.

가장 흔한 회귀식은:

$$
Y_{d,t}=\alpha + \sum_{d\neq d_0}\gamma_d \cdot 1[d] + \sum_{t\neq t_0}\delta_t \cdot 1[t] + \beta \cdot (Post_t \times Treat_d) + \varepsilon_{d,t}
$$

* $d$: 소득분위(1-10)
* $t$: 연도(2006-2025)
* $1[d]$: 분위 고정효과(분위별로 원래 출산율 수준이 다른 걸 제거)
* $1[t]$: 연도 고정효과(모든 분위에 공통인 연도 충격: 경기, 코로나, 금리 등 제거)
* $Post_t$: 2024 이후면 1
* $Treat_d$: “정책 효과가 나타날 것으로 설정한 분위”에 1 (예: 상위 30%면 8-10 분위)

$\beta$가 바로 DID 추정치다.  
“정책 이후, 처리(상위) 분위가 비교(하위/중위) 대비 출산율이 얼마나 더 변했는지”를 준다.

---

#### 1.3 왜 그냥 2023→2025 비교하면 안 되고 DID가 필요한가?

2024-2025에 출산율이 상위에서 올랐다고 해도, 같은 기간에:

* 상위 분위가 더 많이 결혼했을 수도 있고
* 상위 분위가 더 많이 이사/주택구입 했을 수도 있고
* 상위 분위의 30대 비중이 우연히 늘었을 수도 있고
* 고소득 일자리/임금 증가가 상위에 집중됐을 수도 있다

DID는 최소한 “연도 공통 충격”과 “분위 고정 차이”를 제거하고, 남는 추가 변화가 있는지를 보는 장치다.

---

#### 1.4 DID가 성립하려면 필요한 가정: 평행추세(parallel trends)

핵심 가정: 정책이 없었더라도, 처리집단과 비교집단의 결과는 **정책 전에는 비슷한 추세로 움직였을 것**.

이걸 확인하는 대표적 방법이 **이벤트 스터디(event study)** 다.

$$
Y_{d,t}=\alpha + \gamma_d + \delta_t + \sum_{k\neq -1}\beta_k \cdot (1[t - t_0 = k]\times Treat_d)+\varepsilon_{d,t}
$$

* $t_0$: 정책 시작 연도(2024)
* $k$: 정책 시작 대비 상대연도(-3,-2,-1,0,1,2…)

여기서 중요한 건 **정책 전 구간($k<0$)의 $\beta_k$들이 0에 가깝냐**다.  
정책 전부터 이미 상위 분위가 상승추세였다면, 정책 이후 상승을 정책 효과로 볼 수 없다.

---

#### 1.5 실무 설계 포인트(중요)

* **Post를 2024=1로 둘지**, 2024는 시행 초기/월별로 달라서 2025부터 post로 둘지(“실제 노출” 기준) 고민
* 출산율은 **시차**가 있다(임신-출산). 2024 정책이면 출산은 2025에 더 반영될 수도 있다
* 단위가 개인이면: 출산 여부(0/1), 출생아수(카운트), 첫째 출산 hazard 등 다양한 모델이 가능
* 단위가 분위×연도 집계면: 표본이 작아져서(10×20=200) 표준오차/추정 안정성이 떨어질 수 있음

### 2. 정책 수혜 가능 집단 vs 비수혜 집단 비교 — “실제 처치 정의”를 더 정확히

여기서 핵심은: “소득 상위 30%”는 **정책 수혜 가능성**과 1:1로 매칭되지 않는다.

신생아 특례대출은 단순히 소득만이 아니라 (일반적으로) 대출·주택·신용·가구 조건 같은 자격이 붙는다. 그러면 “정책에 노출된 사람”을 더 정확히 정의해야 한다.

---

#### 2.1 왜 ‘수혜 가능성(eligibility)’ 기반 비교가 중요한가?

정책 효과를 보려면 처리집단은 “정책을 실제로 이용했거나 최소한 이용할 수 있었던 집단”이어야 한다.

예를 들어:

* 상위 소득이지만 이미 집을 충분히 갖고 있어 대출 수요가 없으면 효과가 작을 수 있음
* 중위 소득이라도 무주택·신혼·자녀계획이 뚜렷하면 수요가 클 수 있음
* 신용/부채 제약으로 저소득층은 “제도는 있어도 이용이 어렵다”가 될 수 있음

그래서 **(i) 자격요건을 충족하는 집단(eligible)** vs **(ii) 애초에 자격 밖인 집단(ineligible)** 으로 비교하는 게 식별력이 훨씬 좋다.

---

#### 2.2 가장 깔끔한 설계: 소득 컷오프를 이용한 RDD/DiD-RDD

네가 말한 “연소득 1.3억→2억으로 완화(2024년 12월)”는 특히 좋은 신호다. 이런 컷오프는 아래 설계를 가능하게 한다.

* 1.3억 근처 사람들(예: 1.2-1.4억)을 보면, 정책 전후 자격 여부가 급격히 갈림
* 2억 근처도 마찬가지

가능한 접근:

* **RDD(회귀불연속)**: 컷오프 바로 위/아래 비교
* **DiD-RDD**: 컷오프 근처를 정책 전후로 비교(시간 충격 제거)

이게 되면 “상위 30%” 같은 거친 분류보다 훨씬 인과적으로 강하다.

---

#### 2.3 데이터가 제한될 때의 ‘2차선’ 접근: 예측된 수혜확률(Propensity / Exposure)

자격요건이 복잡하고 개별 조건을 다 못 구하면, 정책 도입 전 특성으로 “수혜 가능성”을 추정해서 노출을 연속변수로 쓰는 방법도 있다.

* 예: 무주택/전세거주/주택가격 높은 지역/소득/부채 등으로
* “정책 노출지수”를 만들고

$$
Y_{i,t} = \alpha + \gamma_i + \delta_t + \beta(Post_t \times Exposure_i) + \varepsilon_{i,t}
$$

여기서 $\beta$는 “노출이 1 단위 더 큰 사람일수록 정책 이후 출산이 더 증가했는가”를 본다.

---

#### 2.4 “수혜 가능 vs 비수혜” 비교에서 흔한 함정

* **내생적 선택(selection)**: 실제 대출 이용자는 애초에 출산 의지가 강한 사람일 수 있다  
  → 그래서 “실제 이용자”와 “비이용자” 비교는 위험  
  → 대신 “자격(eligibility)” 또는 “노출(exposure)”을 도구처럼 쓰는 게 낫다
* **정책 동시다발**: 같은 시기 다른 출산/주거 정책도 바뀌면 혼입  
  → 비교집단이 그런 정책에 동일하게 노출되는지 점검 필요

---

### 3. 혼인율, 연령구조, 경기효과 통제 — “상위에서만 오르는 다른 이유”를 제거

DID에 이미 연도 고정효과가 들어가면 “모든 집단에 공통인 경기 충격”은 많이 제거된다. 그런데 문제는:

* 충격이 **소득분위별로 다르게** 작동할 수 있고
* 인구구성 변화(연령, 혼인)가 분위별로 다르게 움직일 수 있다

이걸 통제하지 않으면 $\beta$에 섞여 들어간다(omitted variable bias).

---

#### 3.1 혼인율(또는 혼인상태) 통제

출산은 혼인과 강하게 연결돼 있다(특히 첫째). 상위 분위에서 2024-2025에 혼인이 늘었다면, 출산 증가가 정책 때문이 아니라 혼인 변화 때문일 수 있다.

가능한 변수:

* 분위×연도 혼인율
* 혼인상태별(미혼/기혼) 출산률
* 초혼률, 평균 초혼 연령
* 결혼-출산 시차 고려(혼인 후 1-2년)

회귀에 넣는 방식(집계형 예시):

$$
Y_{d,t}=\alpha+\gamma_d+\delta_t+\beta(Post_t\times Treat_d) + \theta \cdot MarriageRate_{d,t} + \varepsilon_{d,t}
$$

주의: 혼인율이 정책의 결과일 수도 있다(정책이 결혼을 늘려 출산을 늘렸다면). 그 경우 혼인율을 통제하면 “총효과”가 아니라 “직접효과”만 남기게 된다.  
그래서 실무에서는:

* (a) 혼인 통제 없이 총효과 추정
* (b) 혼인 통제 포함해 메커니즘 분해  
  두 개를 같이 보여주는 경우가 많다.

---

#### 3.2 연령구조(특히 20-39 여성 분포) 통제

출산율은 단순히 ‘행동’만이 아니라 ‘구성(composition)’에도 좌우된다.

예: 상위 분위에서 30-34세 비중이 커지면, 동일한 출산행동이어도 TFR이 올라갈 수 있다. 또는 반대로 20대가 줄면 내려간다.

가능한 통제:

* 가임여성(15-49) 중 25-29, 30-34, 35-39 비중
* 평균 출산연령
* 연령표준화(standardization): 연령 구조를 고정해 “행동만” 비교

가장 강한 접근은 **연령특정출산율(ASFR)** 로 쪼개서 분석하는 것:

* $Y_{a,d,t}$: 연령대 $a$의 출산율
* 그러면 연령구조 문제가 상당히 줄어든다(연령대별로 직접 비교하니까)

$$
Y_{a,d,t} = \alpha + \gamma_{a,d} + \delta_t + \beta(Post_t \times Treat_d) + \varepsilon_{a,d,t}
$$

(여기서 $\gamma_{a,d}$는 연령×분위 고정효과)

---

#### 3.3 경기효과 통제: “연도 고정효과만으로 부족한 부분”

연도 고정효과는 전국 공통의 경기 충격을 제거한다. 하지만 상위/하위는 경기 변동에 대한 민감도가 다르다.

예: 상위 분위는 자산가격(주식/부동산)에, 하위 분위는 실업/임금에 더 민감할 수 있다.

그래서 분위×연도 수준에서 아래를 넣기도 한다.

* 분위별 실업률/고용률
* 분위별 실질소득 증가율(임금)
* 자산가격 지표(주택가격, 금리)와 분위 상호작용
* 지역 고정효과까지 붙이면 더 강해짐(지역×연도 충격)

예:

$$
Y_{d,t}=\alpha+\gamma_d+\delta_t+\beta(Post_t\times Treat_d)+\phi X_{d,t}+ \varepsilon_{d,t}
$$

또는

$$
Y_{r,d,t}=\alpha+\gamma_{r,d}+\delta_t+\beta(Post_t\times Treat_d)+\phi X_{r,t}+ \varepsilon_{r,d,t}
$$

(지역 $r$까지 들어가면 식별이 좋아짐)

---

#### 3.4 통제의 목적은 “설명”이 아니라 “식별”

통제를 넣는 이유는 “모형을 그럴듯하게”가 아니라,  
처리효과 $\beta$가 다른 요인(혼인/연령/경기)의 변화에 의해 가짜로 커지거나 작아지는 걸 막기 위해서다.

---

### 4. 이 3개를 합친 ‘권장’ 분석 흐름(실무형)

1. **기초 그림/요약 통계**: 분위별 TFR 추세
2. **이벤트 스터디**로 평행추세 점검(정책 전 계수 0?)
3. **기본 DID**: Post×Treat (Treat는 상위 30% 등)
4. **대체 정의**: Treat를 “자격(eligible)”로 재정의(가능하면 컷오프 활용)
5. **통제 추가**: 혼인/연령/경기 변수 추가하여 $\beta$ 안정성 확인
6. **민감도/위약(placebo)**: 가짜 정책연도(예: 2022)로도 효과가 나오나? 나오면 위험 신호
7. **이질성(heterogeneity)**: 첫째/둘째, 연령대(30-34), 지역(집값 높은 지역) 등

---

원하면, 네가 확보 가능한 데이터 형태(개인 micro인지, 분위×연도 집계인지, 지역까지 있는지)에 맞춰서

* 정확한 회귀식(고정효과 구조)
* 표준오차(클러스터링을 어디에 해야 하는지)
* 이벤트 스터디 플롯을 어떻게 그릴지  
  까지 구체적으로 설계안을 바로 써줄게.

결론부터 말하면, **이 데이터 구조는 ‘기술통계용’으로는 매우 강하고, ‘인과추정용’으로도 잠재력이 크지만, 자동으로 신뢰성이 보장되는 건 아니다.**
“표본이 크다 = 정책 효과가 정확하다”는 전혀 다른 문제다.

아래에서 왜 그런지 **데이터의 장점 → 구조적 한계 → DID 분석에 주는 의미** 순서로 정리한다.

---

### 1. 이 자료가 강력한 이유 (통계적으로 매우 좋은 점)

이건 일반 설문이 아니라 **행정자료 linked dataset**이다. 연구 데이터로서는 상위권 구조다.

#### (1) 출산 측정의 정확도 (➊ 진료내역 DB)

* 병원 진료기록 기반 → 출산 여부 오측정 거의 없음
* 자가보고, 설문 누락 문제 없음
* 출산 시점 정확 → 정책 시점과의 시간 정렬이 가능

👉 결과변수 $Y$ (출산 여부/출산 수)의 **측정오차가 매우 작음**

---

#### (2) 소득 정보의 신뢰성 (➌ 국세청 연계)

* 자가소득 응답이 아니라 **과세자료 기반**
* 특히 고소득 구간 왜곡 적음
* 소득 분위 구분의 정확도 높음

👉 DID에서 중요한 **처리집단 정의(Treat)** 가 신뢰성 높음

---

#### (3) 개인 단위 패널 구조 가능 (➋ 건강보험 가입자격 DB)

* 건강보험 가입자격 정보 = 개인의 고용상태, 직장/지역 가입, 변동 추적 가능
* 개인을 연도별로 추적하는 **패널 구성 가능**

👉 개인 고정효과 모형 가능
→ “원래 출산 성향이 다른 사람들” 문제를 상당 부분 제거 가능

---

#### (4) 가구/세대 구조 파악 가능 (➍ 주민등록 DB)

* 부부 식별 가능
* 세대 분리/결합, 출산 순서 분석 가능
* 가구 소득 기준 분석 가능

👉 단순 여성 개인 분석이 아니라 **가구 단위 출산결정 모델링 가능**

---

#### (5) 표본 규모 (전체의 15%)

이건 그냥 큰 게 아니라:

* 표본 수백만 단위 가능
* 표준오차 작아짐
* 분위 × 연도 × 연령대 같은 세분 분석 가능
* 이벤트 스터디 계수도 안정적으로 추정 가능

즉, **통계적 정밀도(precision)** 는 매우 좋다.

---

### 2. 그런데 왜 이것만으로 “정책 효과 신뢰”가 되지 않는가

여기서 많은 사람들이 착각하는 지점.

> 큰 행정자료는 “측정오차”는 줄여주지만
> “인과 편의(bias)”는 자동으로 제거하지 않는다

#### 핵심 문제는 **식별(identification)** 이다.

행정자료가 아무리 좋아도 아래가 해결되지 않으면 정책 효과는 왜곡될 수 있다.

---

#### (1) 정책 노출의 내생성 (selection bias)

신생아 특례대출은 **대출을 이용할 수 있는 가구**에게 의미가 있다.

그런데:

* 대출 신청하는 가구 = 애초에 출산 계획 있는 가구일 가능성 높음
* 주택 구매 준비된 가구 = 소득 안정적, 혼인 안정적

이 경우

$$
\text{대출 이용} \rightarrow \text{출산 증가}
$$

가 아니라

$$
\text{출산 계획 있음} \rightarrow \text{대출 이용}
$$

일 수 있다.

데이터가 아무리 커도 이 문제는 남는다.

---

#### (2) 동시 변화(confounding shocks)

2024-2025에 동시에:

* 금리 변화
* 주택시장 변화
* 혼인율 반등
* 코로나 후행효과

이런 것들이 상위 소득층에 더 강하게 작용했을 수 있다.
행정자료라도 이걸 통제 안 하면 정책 효과로 오인된다.

---

#### (3) 출산은 구조적으로 ‘지연 변수’

임신 → 출산까지 9-10개월.
정책 시행 시점과 출산 발생 시점이 어긋난다.
이 시차 구조를 고려 안 하면 효과 시점이 왜곡된다.

---

#### (4) 대규모 표본의 착시

표본이 크면:

* 작은 효과도 유의하게 나옴
* 그러나 **유의성(significance) ≠ 인과성(causality)**

오히려 큰 표본은 **아주 작은 편의(bias)도 통계적으로 강하게 나타나게 만든다.**

---

### 3. 이 데이터가 DID 분석에 주는 “진짜 의미”

이 데이터는 **좋은 연구를 할 수 있는 조건**은 갖췄다.
하지만 그건 “방법을 제대로 썼을 때” 이야기다.

이 자료로 가능해지는 고급 설계:

| 가능 분석                | 이유                   |
| -------------------- | -------------------- |
| 개인 고정효과 DID          | 동일 개인의 정책 전후 출산 변화   |
| 부부 단위 분석             | 가구 소득 기준 정책 효과 추정 가능 |
| 연령대별 출산 위험모형(hazard) | 첫째 출산, 둘째 출산 분리 가능   |
| 컷오프 기반 RDD           | 소득요건 완화 시점 활용 가능     |
| 이벤트 스터디              | 정책 전 추세 점검 가능        |

이게 다 되면 → 신뢰도 높은 인과 추정 가능

---

### 4. 그래서 “객관성과 신뢰성” 평가는 이렇게 해야 정확

| 항목        | 평가           |
| --------- | ------------ |
| 출산 측정 정확도 | 매우 높음        |
| 소득 측정 정확도 | 매우 높음        |
| 표본 규모     | 매우 큼         |
| 기술통계 신뢰성  | 매우 높음        |
| 인과추정 신뢰성  | 연구 설계에 따라 다름 |

즉,

> “데이터는 최고급인데, 분석 설계가 인과적으로 엄격하지 않으면 결과는 여전히 상관관계 수준일 수 있다”

가 정확한 평가다.

하지만 **‘조준(DID, 식별전략)’을 제대로 안 하면 명중(인과결론)은 보장되지 않는다.**

## <span style="color: blue;">**4. 데이터 구조에 맞는 분석 설계**</span>

## **ChatGPT 설계**

아래는 네가 적어준 국민건강보험공단 행정자료(진료 ➊, 자격 ➋, 소득 ➌, 세대 ➍) 구조에 딱 맞게 인과추정을 설계한 안이다. 핵심은 (i) 정책 노출을 최대한 정확히 정의하고 (ii) 출산의 시차를 반영하며 (iii) 혼인-연령-경기 같은 교란을 흡수하는 고정효과 구조를 쓰는 거다.

---

### 1. 데이터 구성과 관측 단위

#### 1.1 관측 단위

가장 좋은 단위는 개인-월(person-month) 패널이다. 이유는 정책이 2024년 도입이고, 출산은 월 단위 시차(임신-출산)가 크기 때문이야. 연 단위로 가면 효과 타이밍이 뭉개지고, 2025년 6월까지라는 장점도 반감된다.

- 표본: 2006.01월 - 2025.06월
- 대상: 가임기 여성(예: 15-49세) 중심, 또는 부부 단위로 확장
- 패널: 개인 $i$를 월 $t$로 추적

---

#### 1.2 결과변수(출산 측정)

진료내역 ➊로 출산 이벤트를 잡는다.

- $Birth_{i,t}$: 월 $t$에 출산(분만) 발생이면 1, 아니면 0
- 파생
  - 첫째 출산만: 이전 출산 이력 없는 상태에서 발생한 출산
  - 둘째-셋째 분리: parity별로 별도 분석(정책 반응이 다를 수 있음)
  - 출산시점 정확도: 분만일(또는 월) 기준

---

#### 1.3 소득 변수(처리집단 정의에 핵심)

국세청 연계 소득 ➌를 부부합산 연소득으로 만들 수 있으면 가장 좋다.

- 개인 소득 $Inc_{i,year}$
- 배우자 소득과 합산해 $HHInc_{h,year}$ (세대 DB ➍ + 자격 DB ➋로 배우자 링크)
- 분위: $Decile_{h,year}$ 또는 연속소득
- 정책의 소득요건
  - 초기: $HHInc \le 1.3\text{억}$
  - 2024.12월 이후: $HHInc \le 2\text{억}$
    이 변화는 식별에 엄청 유리하다.

---

#### 1.4 혼인-세대 구조

주민등록 ➍ + 자격 ➋로 아래를 만든다.

- 혼인 상태(가능한 범위에서): 동일 세대 내 배우자 존재, 세대 구성 변화 등으로 근사
- 동거-세대분리: 주거 이벤트로도 사용 가능(정책이 주거금융이라서)
- 출산 전후 세대 이동: 출산과 동시 발생하는 세대변화가 있을 수 있어 타이밍 주의

---

#### 1.5 기타 통제변수 후보(가능하면)

자격 ➋에서

- 고용 형태(직장-지역), 보험료 부과, 자격 변동
- 직장가입자 여부는 경기-안정성의 강한 프록시

지역이 있으면(주소-행정구역)

- 지역×월 고정효과로 지역 경기-주택시장 충격 흡수 가능

---

### 2. 처치(treatment) 정의: 실제 노출을 2단으로 잡기

이 정책은 실제 대출 이용 자료가 없으면 실제 이용자를 처치로 두면 selection이 너무 크다. 그래서 보통은 eligibility(수혜 가능성) 기반으로 간다.

#### 2.1 레벨 A: 정적 eligibility(정책 도입 기준)

- $Eligible^{A}_{h,t}=1$ if $HHInc_{h,\mathrm{year}(t)} \le 1.3\text{억}$ (정책 초기 기준)
- Post: $Post^{A}_t=1$ if $t \ge 2024.01월$

---

#### 2.2 레벨 B: 소득요건 완화 eligibility(2024.12월 변경 활용)

이게 진짜 강한 식별 포인트다.

- $Eligible^{B}_{h,t}=1$ if $HHInc_{h,\mathrm{year}(t)} \le 2\text{억}$
- 새로 편입된 집단 정의

  $$
  NewlyEligible_h = 1\big[1.3\text{억} < HHInc_{h,year} \le 2\text{억}\big]
  $$
  
- 완화 이후

  $$
  Post^{B}_t = 1[t \ge 2024.12\text{월}]
  $$
  
  이렇게 하면 완화 직후에 정책 노출이 갑자기 생긴 집단이 생겨서, DID-RD 설계가 강해진다.

---

### 3. 기본 분석 1: 개인 고정효과 이벤트 스터디 DID(정책 도입 효과)

목표: 2024년 정책 도입 이후, eligible 집단에서 출산이 상대적으로 증가했는지.

#### 3.1 개인-월 선형확률모형(LPM: Linear Probability Model)

$$
Birth_{i,t}=\alpha_i+\delta_t+\beta\,(Post^{A}_t \times Eligible^{A}_{h(i),t}) + \Gamma'X_{i,t} + \varepsilon_{i,t}
$$

- $\alpha_i$: 개인 고정효과(시간불변 출산성향 제거)
- $\delta_t$: 월 고정효과(전국 공통 충격: 금리, 코로나, 정책다발 등 흡수)
- $X_{i,t}$: 시간변동 통제(혼인-세대, 고용자격, 연령대 등)

이 모형의 $\beta$는 정책 도입 이후 eligible 집단의 출산 확률이 non-eligible 대비 얼마나 더 증가했는지를 뜻한다.

---

#### 3.2 이벤트 스터디(평행추세 점검 + 시차 반영)

출산은 시차가 크니 단순 Post 하나로 끝내면 안 되고, 상대시간 계수로 본다.

$$
Birth_{i,t}=\alpha_i+\delta_t+\sum_{k\neq -1}\beta_k\,(1[\mathrm{event\_time}(t)=k]\times Eligible^{A}_{h(i),t})+\Gamma'X_{i,t}+\varepsilon_{i,t}
$$

- event_time은 2024.01월을 0으로 둔 상대월(예: -24 - +18)
- 정책 전 구간($k<0$)에서 $\beta_k$가 0 근처면 평행추세가 그럴듯
- 정책 후 구간($k>0$)에서 언제부터 효과가 뜨는지 확인 가능
- 현실적으로 출산은 최소 9-12개월 시차가 있으니, $k=9$ - $k=18$ 구간에서 반응이 강하게 나오는지 보는 게 핵심

---

#### 3.3 표준오차

- 최소: 개인 클러스터(패널 상관)
- 가능하면: 세대-가구 클러스터(부부-형제 등 상관)도 고려  
  실무에선 보통 개인 클러스터 + (큰 집계 충격이 우려되면) 월 레벨 충격을 월 고정효과로 흡수

---

### 4. 핵심 분석 2: 2024.12월 소득요건 완화에 대한 DID(가장 강한 식별)

이건 정책이 진짜로 처치 강도를 바꾼 시점이라서 식별력이 훨씬 좋아진다. 특히 새로 편입된 1.3 - 2.0억 구간이 자연 처치집단이다.

#### 4.1 DID 모형

$$
Birth_{i,t}=\alpha_i+\delta_t+\beta\,(Post^{B}_t \times NewlyEligible_{h(i)})+\Gamma'X_{i,t}+\varepsilon_{i,t}
$$

- 처치집단: $NewlyEligible=1$ (1.3억 초과 - 2억 이하)
- 비교집단(권장): 컷오프 근처의 약간 아래 집단
  - 예: $HHInc \in [1.1\text{억},1.3\text{억}]$ 또는 $[1.0\text{억},1.3\text{억}]$  
    이렇게 유사한 소득대로 비교하면 교란이 줄어든다.

---

#### 4.2 설계의 장점

- 상위 30% vs 하위 30% 같은 거친 비교보다 훨씬 설득력 있음
- 정책 충격이 2024.12월에 명확히 걸려서 이벤트 타이밍도 명확
- 출산 시차를 고려하면 2025년 상반기에서 효과가 보일 가능성이 높다(정책-임신-출산 라그)

---

#### 4.3 이벤트 스터디 확장

- 2024.12월 기준 상대월로 $\beta_k$를 추정해서, 2025.06월까지의 동학을 본다

---

### 5. 확장 설계: triple-diff(삼중차분)로 교란 더 줄이기

정책이 주거금융 성격이라면, 효과는 무주택-전세-주거불안정 집단에서 더 클 가능성이 크다. 만약 세대-자격 DB로 주거형태를 직접 알기 어렵다면, 근사 변수를 만들 수 있다(예: 세대 이동 잦음, 청년층, 지역, 직장가입 여부 등).

가능하면 주거 관련 프록시 $Z_i$를 두고

$$
Birth_{i,t}=\alpha_i+\delta_t+\beta\,(Post_t\times Eligible_i) + \theta\,(Post_t\times Eligible_i \times Z_i)+\Gamma'X_{i,t}+\varepsilon_{i,t}
$$

여기서 $\theta$가 주거제약이 큰 집단에서 정책 효과가 더 큰가를 준다. 이건 인과설명력을 크게 올려준다(메커니즘과 정합성).

---

### 6. 혼인율-연령구조-경기효과 통제: 이 데이터에서 어떻게 구현할지

#### 6.1 혼인-동거 통제

가능한 구현 순서(데이터 가능성에 따라)

- (가능하면) 혼인 상태 직접변수: 혼인-이혼 행정기록이 있으면 최고
- (근사) 동일 세대 내 배우자 존재, 부부 세대 구성 여부
- (근사) 혼인 전환 이벤트: 어떤 월에 배우자와 같은 세대로 결합되는 시점

통제 방식

- $X_{i,t}$에 혼인상태-동거상태를 넣는 버전
- 또는 표본을 기혼-동거 집단으로 제한해(조건부 효과) 출산 결정만 보기도 함

주의: 혼인은 정책의 경로일 수 있다. 그래서 보통은 두 세트를 같이 보고서에 넣는다.

- 혼인 통제 없이: 총효과
- 혼인 통제 포함: 혼인을 통한 경로를 제거한 직접효과(또는 잔여효과)

---

#### 6.2 연령구조 통제

개인 패널이면 연령구조는 통제하기 쉽다.

- 연령(또는 연령대) 더미를 $X_{i,t}$에 포함
- 더 강하게: 연령×월 고정효과까지도 가능(표본이 크면 가능)

권장: 연령특정 분석

- 25-29, 30-34, 35-39로 쪼개서 별도 이벤트 스터디  
  출산 반응은 거의 항상 30대 초반에서 가장 민감하게 나온다.

---

#### 6.3 경기효과 통제

월 고정효과 $\delta_t$가 전국 공통 경기충격은 거의 흡수한다. 문제는 소득집단별로 충격이 다르게 작동하는 경우다.

가능한 보강

- 직장가입자 여부×월 상호작용(소득-고용안정 충격의 이질성 흡수)
- 지역이 있으면 지역×월 고정효과(주택-경기 지역충격 제거)
- 보험료 변동(소득충격 프록시)을 $X_{i,t}$로 포함

---

### 7. 출산의 시차를 반영한 두 가지 구현 옵션

#### 7.1 옵션 A: 출산 시점 기준(가장 단순)

앞에서 제시한 이벤트 스터디로 정책 이후 몇 개월 뒤 출산이 늘었는지를 직접 본다.

---

#### 7.2 옵션 B: 임신 시작(수태) 근사로 정렬(더 정교)

출산월에서 9개월을 뺀 임신 시작월을 근사해서 정책 노출을 재정렬한다.

- 예: $ConceptionMonth = BirthMonth - 9$  
  이렇게 하면 정책이 임신 결정에 영향을 주는지 더 직접적으로 본다. 데이터가 크면 이 방식이 설득력이 크게 오른다.

---

### 8. 표본 설계(권장)

정책의 타깃은 신생아-출산 가구이니, 아래를 주 분석으로 권장한다.

- 가임여성 20-44세(너무 넓히면 노이즈)
- 과거 출산 이력으로 parity 분리
  - 첫째 출산: 혼인-주거 제약과 강하게 연결
  - 둘째 이상: 소득-주거 안정성과 연결
- 2022.01월 - 2025.06월을 중심으로 이벤트 스터디(정책 전 2년 + 후 18개월)  
  2006년부터 전기간을 쓰는 건 장점도 있지만, 정책 전 추세 검증엔 충분한 길이만 확보하면 된다. 대신 장기 추세는 보조자료로 제시.

---

### 9. 위약(placebo)과 강건성 체크(이걸 해야 신뢰를 주장할 수 있음)

#### 9.1 가짜 정책 시점 위약

- 정책을 2022.01월에 시작한 것처럼 놓고 같은 DID를 돌려본다. 효과가 나오면 설계가 교란을 잡지 못하는 거다.

#### 9.2 비교집단 바꾸기

- eligible 기준을 약간 바꾸거나(1.2억, 1.4억)
- 컷오프 근처 밴드폭을 바꿔도 결과가 비슷한지 확인

#### 9.3 사전 계수(평행추세) 확인

이벤트 스터디에서 정책 전 $\beta_k$들이 체계적으로 0이 아니면 해석을 보수적으로 해야 한다.

---

### 10. 최종 산출물(보고서-논문에 들어갈 결과 세트)

- 도표 1: eligible vs non-eligible의 출산율 추세(월 단위)
- 도표 2: 이벤트 스터디 계수(2024.01월 기준)
- 도표 3: 완화 이벤트 스터디 계수(2024.12월 기준, NewlyEligible 중심)
- 표 1: 기본 DID 추정치(통제 없음-혼인통제-연령통제-경기통제 단계별)
- 표 2: 이질성(연령대, parity, 직장가입 여부, 지역 등)
- 위약 결과 표

---

### 11. 설계 확정에 필요한 추가 정보(체크리스트)

질문을 다시 하라는 뜻이 아니라, 네가 지금 가진 자료에서 가능한 범위를 기준으로 설계 레벨을 확정하려는 체크리스트야.

- 소득이 연 단위인지 월 단위인지(대부분 연 단위일 가능성이 큼)
- 부부 링크가 가능한지(세대-배우자 식별)

#### 2-5. 좌표계 (Coordinate System)

StatsBomb는 **120 × 80 yards** 기준 좌표계를 사용합니다. 공격 방향은 항상 왼쪽에서 오른쪽(x: 0→120)이며, y축은 0(top)에서 80(bottom)입니다. 슈팅의 `end_location`은 3차원(x, y, z)으로 z는 높이를 나타냅니다.

### 3. 라인업 파일 (Lineup Data)

라인업 파일은 양팀의 전체 엔트리 정보를 담고 있습니다. 프랑스 23명, 크로아티아 22명으로 총 45명의 선수 데이터가 포함되어 있습니다.

#### 3-1. 필드 구조

| 필드명 | 설명 |
|--------|------|
| `team_id` / `team_name` | 팀 식별 정보 |
| `player_id` / `player_name` | 선수 식별 정보 |
| `player_nickname` | 선수 별명 (예: Kylian Mbappé) |

| `positions` | 포지션 변경 이력 (시간, 사유 포함) |

### 3-2. 포지션 변경 이력 (positions)

`positions` 배열은 경기 중 선수의 포지션 변경을 시간순으로 기록합니다. 각 항목에는 `position`(포지션명), `from`/`to`(시간), `from_period`/`to_period`(반), `start_reason`/`end_reason`(변경 사유: Starting XI, Substitution, Tactical Shift, Final Whistle 등)이 포함됩니다.

**예시:** Kylian Mbappé는 Right Midfield로 선발 출전한 후, 81분에 Tactical Shift로 Right Center Forward로 변경되어 경기 종료까지 해당 포지션을 유지했습니다.

### 4. 활용 방안 (Potential Use Cases)

- **경기 흐름 분석:** `timestamp`와 `possession` 시퀀스를 활용하여 경기 흐름을 시각화
- **패스 네트워크:** Pass 이벤트의 passer→recipient 관계로 패스 네트워크 구성
- **xG 모델 검증:** Shot 이벤트의 `statsbomb_xg`와 `freeze_frame`을 활용한 기대골 분석
- **수비 분석:** Pressure, Duel, Interception, Ball Recovery 데이터로 수비 강도 측정
- **히트맵 / 시각화:** `location` 좌표를 활용한 슛, 패스, 압박 등의 공간 분석
- **선수 평가:** Lineup의 positions 변경 이력 + 이벤트 데이터를 결합한 개인 퍼포먼스 분석

### 5. 참고 사항 (Notes)

- **데이터 출처:** StatsBomb Open Data (https://github.com/statsbomb/open-data)
- `Ball Receipt*`의 `*`는 해당 이벤트가 불완전한 수신(miscontrol)일 수 있음을 나타냅니다.
- `under_pressure` (Boolean)는 상대 압박 하에 수행된 이벤트에 추가로 표시됩니다.
- `related_events`를 통해 이벤트 간 연결 관계를 파악할 수 있습니다 (예: 패스 ↔ 수신, 슈팅 ↔ 세이브).
- **xG(기대골):** 프랑스 1.10, 크로아티아 1.48 — 크로아티아가 xG에서 앞섰으나 실제 결과는 역전.

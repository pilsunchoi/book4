## **Claude 설명**

이 그래프는 **구매력평가(PPP, Purchasing Power Parity)** 의 예측을 실증적으로 검증한 산점도입니다.

**축 설명:**

- **가로축**: 1975~2005년 동안 각국의 인플레이션율과 미국 인플레이션율의 차이 (연간 %)
- **세로축**: 같은 기간 동안 각국 통화의 대미 달러 감가상각률 (연간 %)

**핵심 내용:**

검은 대각선은 PPP가 예측하는 이론적 관계, 즉 "인플레이션 차이만큼 환율이 변동한다"는 것을 나타냅니다. 예를 들어 어떤 나라의 인플레이션이 미국보다 연 10% 높으면, 그 나라 통화는 연 10% 감가상각될 것이라는 예측입니다.

각 점은 개별 국가를 나타내며, 대부분의 점이 이 대각선 근처에 분포하고 있습니다. 이는 장기적으로 PPP가 꽤 잘 성립한다는 것을 보여줍니다.

**해석:**

- 인플레이션이 낮은 나라(왼쪽)는 통화 감가상각도 적고
- 인플레이션이 높은 나라(오른쪽, +30~40%)는 통화 가치도 크게 하락했습니다
- 단기적으로는 PPP에서 벗어나는 경우가 많지만, 30년이라는 긴 기간에 걸쳐 보면 인플레이션 차이와 환율 변동 사이에 강한 양의 상관관계가 나타납니다

국제경제학 교과서에서 PPP의 장기적 유효성을 설명할 때 자주 사용되는 대표적인 그래프입니다.

## <span style="color: blue;">**감가상각률이 아니라 절하율이다!**</span>

## **Claude 설명**

맞습니다, 수정합니다. "depreciation"을 여기서는 **통화의 절하율**로 번역하는 것이 더 적절합니다.

세로축은 각국 통화의 대미 달러 **절하율**(연간 %)을 나타냅니다. 즉, 인플레이션이 미국보다 높은 나라일수록 해당 국가의 통화가 달러 대비 더 많이 절하되었다는 것을 보여주는 그래프입니다.

나머지 해석은 동일합니다. 장기적으로 인플레이션 차이만큼 환율이 절하된다는 PPP의 예측이 실증적으로 잘 들어맞는다는 것을 보여주고 있습니다.

## **ChatGPT 설명**

이 그림은 구매력평가설(PPP, Purchasing Power Parity)의 장기적 예측이 실제 데이터와 얼마나 일치하는지를 보여주는 산점도이다.

### 1. 축의 의미

**가로축**  
1975–2005년 동안의 인플레이션 차이 $\pi_{country} - \pi_{US}$ (미국보다 물가상승률이 얼마나 높았는지, 연평균 %)

**세로축**  
같은 기간 동안의 환율 절하율 (해당 국가 통화가 달러 대비 얼마나 약해졌는지, 연평균 %)

즉,

- 가로축: 물가 차이  
- 세로축: 환율 변화  

---

### 2. PPP가 예측하는 관계

구매력평가설의 기본식은

$$
\Delta s \approx \pi_{domestic} - \pi_{foreign}
$$

여기서

- $\Delta s$: 환율 상승률(자국 통화 절하율)  
- $\pi$: 인플레이션율  

의미는 다음과 같다.

자국 물가가 미국보다 더 빨리 오르면 → 자국 통화는 그만큼 절하된다.

그래프의 대각선(45도 선)이 바로 PPP가 예측하는 관계이다.

---

### 3. 점들의 의미

각 점은 한 국가를 나타낸다.

관찰되는 특징:

- 대부분의 점이 대각선 주변에 밀집  
- 인플레이션 차이가 클수록 환율 절하도 큼  

즉, 장기적으로 PPP가 꽤 잘 맞는다.

예:

인플레이션이 미국보다 +10%p 높으면  
→ 통화도 약 +10% 정도 절하되는 경향

---

### 4. 경제적 해석

이 그림이 말하는 핵심은 다음이다.

장기적으로 환율은 물가 차이를 반영한다.

즉,

- 고인플레이션 국가 → 통화 약세  
- 저인플레이션 국가 → 통화 강세 또는 안정  

이는 국제경제에서 매우 중요한 장기 관계이다.

---

### 5. 중요한 추가 해석 (실증 관점)

- 단기에서는 PPP가 잘 맞지 않음 (환율 변동성 큼)  
- 그러나 30년 같은 장기 평균에서는 강한 선형 관계  

완전히 일치하지 않는 이유:

- 자본 이동  
- 생산성 차이 (Balassa–Samuelson 효과)  
- 정책, 위기, 금융시장 요인  

---

### 요약

장기적으로

$$
\text{환율 변화} \approx \text{인플레이션 차이}
$$

이 그래프는 이러한 구매력평가설이 데이터에서 상당히 잘 지지된다는 것을 보여준다.


## <span style="color: blue;">**2. 그래프 기간 업데이트**</span>

## **ChatGPT 분석**

아래는 **업데이트된 PPP 산점도**를 만들 때 사용한 **데이터 출처**와 **변수 정의**를 정확히 정리한 것이다. 핵심은 World Bank의 장기 국가별 패널 데이터를 이용해 다음을 계산하는 것이다.

$$\text{평균 인플레이션 격차} \quad \text{vs} \quad \text{평균 환율 절하율}$$

---

### 1. 데이터 출처

#### World Bank: World Development Indicators (WDI)

* 제공 기관: World Bank
* 데이터베이스: **World Development Indicators**
* URL
  [https://data.worldbank.org/](https://data.worldbank.org/)
* API 사용:

  ```
  https://api.worldbank.org/v2/country/all/indicator/{indicator}
  ```

특징:

* 거의 모든 국가 포함
* 연도별 시계열 (1960–최근)
* 국제 비교 연구에서 가장 표준적인 공개 데이터

---

### 2. 사용 변수

#### (1) 인플레이션

**지표 코드**

```
FP.CPI.TOTL.ZG
```

**이름**
Inflation, consumer prices (annual %)

**정의**
$
\pi_{i,t} = \frac{CPI_{i,t} - CPI_{i,t-1}}{CPI_{i,t-1}} \times 100
$

의미:

* 국가 $i$의 소비자물가지수(CPI) 기준 연간 물가상승률 (%)
* PPP 이론에서 사용하는 **상대 가격 변화의 근사치**

---

#### (2) 환율

**지표 코드**

```
PA.NUS.FCRF
```

**이름**
Official exchange rate (LCU per US$, period average)

**정의**

환율:
$
e_{i,t} = \text{local currency units per US dollar}
$

예:

* 한국: 원/달러
* 일본: 엔/달러

해석:

* $e$ 상승 → 자국통화 **절하**
* $e$ 하락 → 자국통화 **절상**

---

### 3. PPP 분석을 위한 파생 변수

#### (1) 인플레이션 격차 (미국 대비)

연도별:
$
d_{i,t} = \pi_{i,t} - \pi_{US,t}
$

기간 평균:
$
\overline{d_i} = \frac{1}{T} \sum_t d_{i,t}
$

그래프 **가로축**

---

#### (2) 평균 환율 절하율

로그 변화율 사용:
$
\Delta s_{i,t} = \ln(e_{i,t}) - \ln(e_{i,t-1})
$

기간 평균:
$
\overline{\Delta s}_i = \frac{1}{T} \left[\ln(e_{i,T}) - \ln(e_{i,0})\right]
$

퍼센트로 표시:
$
100 \times \overline{\Delta s}_i
$

그래프 **세로축**

---

### 4. PPP 이론과의 연결

상대적 PPP:
$
\Delta s_i \approx \pi_i - \pi_{US}
$

따라서 산점도에서

* x축: 평균 인플레이션 격차
* y축: 평균 환율 절하율
* 기준선:
  $
  y = x
  $

---

### 5. 왜 이 변수 조합을 쓰는가 (실증적 이유)

1. CPI 인플레이션
   → 국가 간 가격수준 변화의 표준 지표

2. LCU/USD 환율
   → 절하 방향이 PPP 식과 일치

3. 로그 변화율
   → 장기 성장률(연평균) 해석 가능

4. 기간 평균
   → 단기 환율 잡음을 제거하고 **장기 PPP 관계** 확인

---

### 6. 해석 예시

어떤 국가에서

* 평균 인플레이션 격차: +8%
* 평균 절하율: +9%

→ PPP가 장기적으로 잘 성립

---

### 7. 참고 (원그림과의 차이)

원그림:

* 기간: 1975–2005
* 일부 국가 샘플

업데이트 버전:

````python
# 그래프 기간 업데이트

import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

INF = "FP.CPI.TOTL.ZG"   # Inflation, consumer prices (annual %)
EXR = "PA.NUS.FCRF"      # Official exchange rate (LCU per US$, period average)

def wb_indicator_all_countries(indicator: str, start: int, end: int) -> pd.DataFrame:
    # World Bank API: all countries, JSON
    url = f"https://api.worldbank.org/v2/country/all/indicator/{indicator}"
    params = {
        "format": "json",
        "per_page": 20000,
        "date": f"{start}:{end}",
    }

    r = requests.get(url, params=params, timeout=60)
    r.raise_for_status()
    data = r.json()
    if not isinstance(data, list) or len(data) < 2:
        raise RuntimeError("Unexpected World Bank API response")

    rows = []
    for obs in data[1]:
        # obs: {"country": {"id","value"}, "date": "YYYY", "value": ...}
        if obs.get("value") is None:
            continue
        rows.append({
            "country_code": obs["country"]["id"],
            "country": obs["country"]["value"],
            "year": int(obs["date"]),
            "value": float(obs["value"]),
        })
    return pd.DataFrame(rows)

def build_ppp_scatter(start=1975, end=2024, min_years=20, drop_extreme_inflation=200.0):
    # 1) 데이터 로드
    inf = wb_indicator_all_countries(INF, start, end).rename(columns={"value": "infl"})
    exr = wb_indicator_all_countries(EXR, start, end).rename(columns={"value": "exr"})

    # 2) 미국 인플레이션 시계열(연도별)
    us_inf = (inf[inf["country_code"] == "US"][["year", "infl"]]
              .rename(columns={"infl": "infl_us"}))

    # 3) 인플레이션 격차: (국가 - 미국), 연도별로 맞춘 뒤 평균
    inf2 = inf.merge(us_inf, on="year", how="inner")
    inf2["infl_diff"] = inf2["infl"] - inf2["infl_us"]

    # 너무 극단적인 인플레이션(예: 초고인플레) 제거 옵션
    inf2 = inf2[inf2["infl"].abs() <= drop_extreme_inflation]

    infl_summary = (inf2.groupby(["country_code", "country"])
                        .agg(infl_diff_mean=("infl_diff", "mean"),
                             n_years_infl=("infl_diff", "count"))
                        .reset_index())

    # 4) 환율 절하율: LCU/USD의 로그변화율 (연평균)
    #    e가 커지면(LCU/USD 상승) 자국통화 절하.
    exr = exr.sort_values(["country_code", "year"])
    exr["ln_exr"] = np.log(exr["exr"])

    exr_summary = (exr.groupby(["country_code", "country"])
                      .agg(year_min=("year", "min"),
                           year_max=("year", "max"),
                           ln_exr_min=("ln_exr", "first"),
                           ln_exr_max=("ln_exr", "last"),
                           n_years_exr=("ln_exr", "count"))
                      .reset_index())

    # 기간이 start~end 전체가 아닐 수 있으니, 최소 관측 연수 필터
    exr_summary["T"] = exr_summary["year_max"] - exr_summary["year_min"]
    exr_summary = exr_summary[exr_summary["T"] >= 1]

    # 연평균 절하율(%) = 100 * (ln e_end - ln e_start) / T
    exr_summary["dep_rate"] = 100.0 * (exr_summary["ln_exr_max"] - exr_summary["ln_exr_min"]) / exr_summary["T"]

    # 5) 결합 + 필터(인플레이션/환율 둘 다 충분히 긴 시계열)
    df = infl_summary.merge(exr_summary, on=["country_code", "country"], how="inner")
    df = df[(df["n_years_infl"] >= min_years) & (df["n_years_exr"] >= min_years)]

    return df

def plot_ppp_scatter(df: pd.DataFrame, title: str):
    x = df["infl_diff_mean"].to_numpy()
    y = df["dep_rate"].to_numpy()

    # 축 범위는 데이터에 맞춰 약간 여유를 둠
    lo = np.nanmin([x.min(), y.min()])
    hi = np.nanmax([x.max(), y.max()])
    pad = 0.05 * (hi - lo) if hi > lo else 1.0
    lo2, hi2 = lo - pad, hi + pad

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, alpha=0.7)
    plt.plot([lo2, hi2], [lo2, hi2])  # PPP 예측선: y = x
    plt.xlim(lo2, hi2)
    plt.ylim(lo2, hi2)
    plt.xlabel("Inflation differential vs U.S. (avg % per year)")
    plt.ylabel("Rate of depreciation vs U.S. dollar (avg % per year)")
    plt.title(title)
    plt.show()

# 실행 예시: 1975~가능한 최신(예: 2024)로 업데이트
df = build_ppp_scatter(start=1975, end=2024, min_years=20, drop_extreme_inflation=200.0)
plot_ppp_scatter(df, "Relative PPP: depreciation vs inflation differential (1975–2024)")
df
````

````python
# 구매력평가 이론 2개 기간 비교

import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

INF = "FP.CPI.TOTL.ZG"   # 인플레이션(CPI 연간 %)
EXR = "PA.NUS.FCRF"      # 공식 환율(현지통화/달러, 연평균)

def wb_indicator_all_countries(indicator: str, start: int, end: int) -> pd.DataFrame:
    url = f"https://api.worldbank.org/v2/country/all/indicator/{indicator}"
    params = {"format": "json", "per_page": 20000, "date": f"{start}:{end}"}
    r = requests.get(url, params=params, timeout=60)
    r.raise_for_status()
    data = r.json()
    if not isinstance(data, list) or len(data) < 2:
        raise RuntimeError("Unexpected World Bank API response")

    rows = []
    for obs in data[1]:
        if obs.get("value") is None:
            continue
        rows.append({
            "country_code": obs["country"]["id"],
            "country": obs["country"]["value"],
            "year": int(obs["date"]),
            "value": float(obs["value"]),
        })
    return pd.DataFrame(rows)

def build_ppp_period(start: int, end: int, min_years: int = 20, drop_extreme_inflation: float = 200.0) -> pd.DataFrame:
    # 1) 데이터 로드
    inf = wb_indicator_all_countries(INF, start, end).rename(columns={"value": "infl"})
    exr = wb_indicator_all_countries(EXR, start, end).rename(columns={"value": "exr"})

    # 2) 미국 인플레이션 (연도별)
    us_inf = (inf[inf["country_code"] == "US"][["year", "infl"]]
              .rename(columns={"infl": "infl_us"}))

    # 3) 인플레이션 격차 (국가 - 미국), 연도별 결합 후 평균
    inf2 = inf.merge(us_inf, on="year", how="inner")
    inf2 = inf2[inf2["infl"].abs() <= drop_extreme_inflation]  # 초고인플레 완화용(옵션)
    inf2["infl_diff"] = inf2["infl"] - inf2["infl_us"]

    infl_summary = (inf2.groupby(["country_code", "country"])
                        .agg(infl_diff_mean=("infl_diff", "mean"),
                             n_years_infl=("infl_diff", "count"))
                        .reset_index())

    # 4) 환율 절하율: ln(e)의 기간 평균 성장률(연평균 %)
    exr = exr.sort_values(["country_code", "year"])
    exr["ln_exr"] = np.log(exr["exr"])

    # 국가별로 가능한 범위 내(해당 기간 내) 첫/끝을 잡아 연평균 변화율 계산
    exr_summary = (exr.groupby(["country_code", "country"])
                      .agg(year_min=("year", "min"),
                           year_max=("year", "max"),
                           ln_exr_min=("ln_exr", "first"),
                           ln_exr_max=("ln_exr", "last"),
                           n_years_exr=("ln_exr", "count"))
                      .reset_index())

    exr_summary["T"] = exr_summary["year_max"] - exr_summary["year_min"]
    exr_summary = exr_summary[exr_summary["T"] >= 1]
    exr_summary["dep_rate"] = 100.0 * (exr_summary["ln_exr_max"] - exr_summary["ln_exr_min"]) / exr_summary["T"]

    # 5) 결합 + 최소 관측치 필터
    df = infl_summary.merge(exr_summary, on=["country_code", "country"], how="inner")
    df = df[(df["n_years_infl"] >= min_years) & (df["n_years_exr"] >= min_years)].copy()

    df["period"] = f"{start}-{end}"
    return df[["country_code","country","period","infl_diff_mean","dep_rate","n_years_infl","n_years_exr","year_min","year_max"]]

def ols_beta(x, y):
    # y = a + b x
    X = np.column_stack([np.ones_like(x), x])
    beta = np.linalg.lstsq(X, y, rcond=None)[0]
    a, b = beta[0], beta[1]
    yhat = a + b*x
    resid = y - yhat
    return a, b, yhat, resid

def period_metrics(df: pd.DataFrame):
    x = df["infl_diff_mean"].to_numpy()
    y = df["dep_rate"].to_numpy()

    a, b, yhat, resid = ols_beta(x, y)
    corr = np.corrcoef(x, y)[0,1]

    # PPP 기준선 y=x 대비 오차
    ppp_err = y - x
    rmse_ppp = np.sqrt(np.mean(ppp_err**2))
    mae_ppp = np.mean(np.abs(ppp_err))

    # 회귀 적합도 (R^2)
    ss_res = np.sum(resid**2)
    ss_tot = np.sum((y - y.mean())**2)
    r2 = 1 - ss_res/ss_tot if ss_tot > 0 else np.nan

    return {
        "N": len(df),
        "corr(x,y)": corr,
        "OLS a": a,
        "OLS b": b,
        "R2": r2,
        "RMSE(y-x)": rmse_ppp,
        "MAE(y-x)": mae_ppp,
    }

def plot_two_periods(df1: pd.DataFrame, df2: pd.DataFrame, title1: str, title2: str):
    def _plot(ax, df, title):
        x = df["infl_diff_mean"].to_numpy()
        y = df["dep_rate"].to_numpy()

        lo = np.nanmin([x.min(), y.min()])
        hi = np.nanmax([x.max(), y.max()])
        pad = 0.05*(hi-lo) if hi > lo else 1.0
        lo2, hi2 = lo - pad, hi + pad

        ax.scatter(x, y, alpha=0.7)
        ax.plot([lo2, hi2], [lo2, hi2])  # PPP: y=x
        ax.set_xlim(lo2, hi2)
        ax.set_ylim(lo2, hi2)
        ax.set_xlabel("Inflation differential vs U.S. (avg % per year)")
        ax.set_ylabel("Depreciation vs US$ (avg % per year)")
        ax.set_title(title)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    _plot(axes[0], df1, title1)
    _plot(axes[1], df2, title2)
    plt.show()

# =========================
# 실행: 두 기간 계산
# =========================
min_years = 20

df_7505 = build_ppp_period(1975, 2005, min_years=min_years)
df_7524 = build_ppp_period(1975, 2024, min_years=min_years)

# =========================
# 동일 국가만 남기기 (교집합)
# =========================
common = sorted(set(df_7505["country_code"]).intersection(set(df_7524["country_code"])))

df_7505_c = df_7505[df_7505["country_code"].isin(common)].copy()
df_7524_c = df_7524[df_7524["country_code"].isin(common)].copy()

print("공통 국가 수:", len(common))

# =========================
# 비교용 요약 지표
# =========================
m1 = period_metrics(df_7505_c)
m2 = period_metrics(df_7524_c)
summary = pd.DataFrame([m1, m2], index=["1975-2005 (common)", "1975-2024 (common)"])
display(summary)

# =========================
# 산점도 비교 (동일 국가 대상)
# =========================
plot_two_periods(
    df_7505_c, df_7524_c,
    title1="Relative PPP (1975–2005) — same countries",
    title2="Relative PPP (1975–2024) — same countries"
)
````

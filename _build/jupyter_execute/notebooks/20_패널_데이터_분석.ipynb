다음은 업로드된 `wage_panel.csv` 데이터의 구조와 변수에 대한 개요이다. 이 데이터는 개인을 여러 시점에 걸쳐 추적한 **패널 데이터(panel data)** 형태로, 임금 결정 요인을 분석할 때 자주 사용된다.

---

### 1. 데이터 구조

이 데이터는 **개인(person)** 을 **여러 연도(year)**에 걸쳐 반복 관측한 **균형 패널(balanced panel)** 이다.

* 관측 단위: 개인–연도 (person-year)
* 패널 식별자:

  * `nr`: 개인 ID
  * `year`: 관측 연도
* 특징:

  * 동일한 개인이 여러 시점에서 반복 등장
  * 개인 고유 특성(능력, 성향 등)과 시간 변화 요인을 동시에 분석 가능

패널 데이터 형태:
$$
i = 1, \dots, N \quad (개인), \quad
t = 1, \dots, T \quad (연도)
$$

---

### 2. 주요 변수 설명

#### (1) 종속변수 (임금 관련)

* `lwage`: 로그 임금

  * 시간당 임금의 로그값
  * 해석: 계수는 대략 **퍼센트 변화** 의미

---

#### (2) 인적자본 변수

* `educ`: 교육 연수 (years of schooling)
* `exper`: 총 노동경험 연수
* `expersq`: 경험의 제곱

  * 경험 효과의 **비선형성**(수익 체감) 반영

---

#### (3) 근속 및 노동시장 상태

* `tenure`: 현재 직장에서의 근속연수
* `tenursq`: 근속연수 제곱
* `union`: 노조 가입 여부 (1 = 가입)
* `hours`: 연간 근로시간

---

#### (4) 인구통계 변수

* `black`: 흑인 여부 (1 = 흑인)
* `hisp`: 히스패닉 여부 (1 = 히스패닉)
* `married`: 결혼 여부 (1 = 기혼)
* `south`: 남부 지역 거주 여부 (1 = 남부)
* `urban`: 도시 거주 여부 (1 = 도시)

---

### 3. 분석에서의 의미

이 데이터는 다음과 같은 계량경제 분석에 적합하다.

#### (1) 임금 결정 모형

$$
\log(wage_{it}) = \beta_0 + \beta_1 educ_i + \beta_2 exper_{it} + \beta_3 tenure_{it} + \alpha_i + u_{it}
$$

* $\alpha_i$: 개인 고유 효과 (능력, 성실성 등 관측 불가 요인)

---

#### (2) 패널 모형 활용 가능

* Pooled OLS
* Fixed Effects (FE)
* Random Effects (RE)
* Hausman 검정

특히,

* `educ`처럼 **시간에 따라 변하지 않는 변수**는 FE에서 제거됨
* `exper`, `tenure`처럼 **시간에 따라 변하는 변수**는 FE에서 식별됨

* 개인 고유 능력을 통제하면 결과가 어떻게 달라지는가?
* FE와 RE 중 어떤 모형이 더 적절한가?

### 4. Hausman 검정 (FE vs RE)

**검정 아이디어**

- 귀무가설 $H_0$: 개체효과와 설명변수가 독립이라서 RE가 일치추정량(consistent)  
- 대립가설 $H_1$: 독립이 아니어서 RE는 불일치, FE가 더 안전  

여기서는 FE와 RE의 공통 계수(`exper`, `expersq`, `union`, `married`)를 대상으로 계산

**결과**

**(clustered 공분산 기준)**  
- 통계량(chi-square) = 16.1453  
- 자유도 df = 4  
- p-value = 0.00283023  

**(non-robust 공분산 기준)**  
- 통계량(chi-square) = 31.5736  
- 자유도 df = 4  
- p-value = 2.33805e-06  

**결론**

- p-value가 매우 작으므로 $H_0$를 기각  
- 이 사양에서는 RE의 독립 가정이 깨졌다고 보는 쪽이 자연스럽고, 실무적으로는 FE를 우선 선택하는 것이 합리적이다.

---

### 5. 한 줄 정리

- pooled는 개인 불변 특성을 통제하지 못해서 `union`, `married` 같은 계수를 크게 잡을 수 있다
- FE로 가면 개인 고유 요인을 제거한 순수한 within 효과를 보게 되고, Hausman도 FE 쪽을 지지한다

## <span style="color: blue;">**4. 패널 데이터 모형의 직관적 이해**</span>

## **ChatGPT 설명**

패널 데이터는 같은 개인(또는 기업, 지역 등)을 여러 시점에 걸쳐 반복 관측한 자료다. 핵심 문제는 이것이다.

>각 개인은 보이지 않는 고유 특성(능력, 성실성, 성향 등)을 가지고 있는데, 이것이 설명변수와 상관될 수 있다. 이 “숨은 개인 차이”를 어떻게 처리하느냐에 따라 모형이 달라진다.

아래를 하나의 예로 생각하자.

* 목표: 임금에 영향을 주는 요인 분석
* 설명변수: 교육, 경험, 노조 여부 등
* 문제: 어떤 사람은 원래 능력이 높다. 이 능력은 관측되지 않지만 임금에도 영향을 주고, 교육 수준과도 관련 있을 수 있다.


이 보이지 않는 개인 특성을 $\alpha_i$라고 두자. 전체 구조는 이렇게 생각하면 된다.

$$
y_{it} = \beta X_{it} + \alpha_i + u_{it}
$$

여기서

* $i$: 개인
* $t$: 시간
* $\alpha_i$: 개인 고유 특성(시간에 따라 변하지 않음)

이 $\alpha_i$를 어떻게 다루느냐가 핵심이다.

---

### 1. 통합 OLS (Pooled OLS)

**아이디어**

모든 데이터를 그냥 하나로 합쳐서 일반 OLS를 돌린다.

즉,

* 개인 구분 무시
* 시간 구분 무시
* 그냥 관측치가 많아진 단순 회귀

**직관**

“모든 사람이 똑같은 출발선에 있다”고 가정하는 것과 같다.

**문제**

만약 개인 능력 같은 $\alpha_i$가

* 임금에 영향 있고
* 교육 같은 변수와도 관련 있다면

→ 설명변수와 오차가 상관됨
→ OLS 편향 발생

**예**

능력 높은 사람일수록 교육도 많이 받고 임금도 높다.
OLS는 교육 효과를 과대평가할 수 있다.

**언제 사용?**

* 개인 차이가 중요하지 않을 때
* 또는 개인 효과가 설명변수와 독립이라고 믿을 때

---

### 2. 고정효과 모형 (Fixed Effects, FE)

**아이디어**

각 개인마다 고유한 기준선을 따로 둔다.

즉,

* 사람마다 평균 수준은 다르다고 인정
* 하지만 그 차이는 시간에 따라 변하지 않는다고 가정

**직관**

“사람 간 비교는 하지 말고, 같은 사람 안에서의 변화만 보자”

**예**

A의 임금 변화 vs A의 경험 변화
B의 임금 변화 vs B의 경험 변화

→ 개인 능력 같은 고정된 특성은 자동으로 제거된다.

수학적으로는 개인 평균을 빼는 방식:

$$
(y_{it} - \bar{y}_i) = \beta (X_{it} - \bar{X}_i) + (u_{it} - \bar{u}_i)
$$

여기서 $\alpha_i$는 사라진다.

**장점**

* 관측되지 않는 개인 특성이 설명변수와 상관되어도 문제 없음
* 인과 해석에 더 안전

**단점**

* 시간에 따라 변하지 않는 변수는 추정 불가
  (예: 성별, 인종, 교육이 변하지 않으면 효과 추정 안 됨)

---

### 3. 확률효과 모형 (Random Effects, RE)

**아이디어**

개인 효과를 제거하지 않고, 하나의 “확률 변수”로 취급한다.

$$
\alpha_i \sim \text{random}
$$

**핵심 가정**

$$
\alpha_i \text{ 와 } X_{it} \text{ 는 독립}
$$

**직관**

“사람 간 차이는 있지만, 그 차이는 설명변수와 관련 없는 순수한 랜덤 잡음이다”

이 가정이 맞으면

* 개인 간 정보 + 개인 내 정보
  둘 다 사용 가능
  → FE보다 효율적 (표준오차 더 작음)

**장점**

* 시간불변 변수도 추정 가능 (예: 성별, 교육)
* 효율적

**단점**

* 만약 $\alpha_i$가 설명변수와 상관되면
  → 심각한 편향 발생

---

### 4. FE vs RE의 직관적 차이

**FE의 생각**

“사람마다 숨은 특성이 있고, 그것이 설명변수와 관련 있을 수 있다. 안전하게 제거하자.”

**RE의 생각**

“사람마다 차이는 있지만, 그건 설명변수와 상관없는 랜덤이다. 제거할 필요 없다.”

**핵심 질문**

개인 효과 $\alpha_i$가 설명변수와 상관되어 있는가?

---

### 5. Hausman 검정

이 질문을 통계적으로 검사하는 방법이다.

**아이디어**

* FE는 항상 일치추정량 (가정이 약함)
* RE는 가정이 맞으면 더 효율적
* 가정이 틀리면 RE는 편향됨

따라서

RE가 맞다면
→ FE와 RE 결과가 비슷해야 한다

**검정 논리**

귀무가설 $H_0$: 
RE가 타당하다
(개인효과와 설명변수가 독립)

대립가설 $H_1$: 
RE가 잘못되었다
(FE를 써야 한다)

**결과 해석**

* p-value 작음 → FE와 RE가 크게 다름 → RE 가정 깨짐 → FE 선택
* p-value 큼 → 차이 없음 → RE 사용 가능

**직관**

“RE를 써도 되는 상황인가?”를 묻는 검정

---

### 6. 한 번에 정리 (직관)

**Pooled OLS**
→ 사람 차이 무시

**Fixed Effects**
→ 사람마다 고유 특성 있음, 그리고 그것이 X와 관련 있을 수 있음 → 제거

**Random Effects**
→ 사람 차이는 있지만 X와 무관한 랜덤 효과 → 효율적으로 사용

**Hausman**
→ RE의 독립 가정이 맞는지 검사
→ 틀리면 FE 사용

---

실무 감각으로 정리하면

* 패널 데이터에서 인과 해석이 목적이면 → FE부터 생각
* 시간불변 변수 효과가 필요하면 → RE 고려
* 그리고 Hausman으로 최종 판단

원하면

“왜 FE에서 계수가 작아지는 경우가 많은지”도 직관적으로 이어서 설명해 줄 수 있다.

```python
# =========================================
# 공통 준비: 데이터 로드 + 패널 인덱스 설정
# =========================================
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.panel import PooledOLS, PanelOLS, RandomEffects
from linearmodels.datasets import wage_panel

# 1) 데이터 로드
data = wage_panel.load()
df = data.copy()

# 2) 패널 인덱스 설정: (개인ID=nr, 연도=year)
df = df.set_index(["nr", "year"]).sort_index()

# 3) (선택) 숫자형 컬럼만 따로 보고 싶으면 아래처럼
# num_cols = df.select_dtypes(include=[np.number]).columns
# df[num_cols].head()
```

```python
# =========================================
# (1) 변수 요약통계
# =========================================

# 식별자(nr, year)는 이미 인덱스라 컬럼에 없다.
# 회귀에 쓸 만한 숫자형 변수들(결측치 포함 가능)을 기준으로 요약통계를 만든다.
# (원하면 특정 변수만 골라서 summary_cols를 구성해도 된다.)

summary_cols = df.columns  # 모든 컬럼 요약
# summary_cols = ["lwage", "educ", "exper", "expersq", "union", "married", "black", "hisp", "hours"]  # 예시

summary = df[summary_cols].describe().T  # count, mean, std, min, 25%, 50%, 75%, max

# 출력
summary
```

```python
# =========================================
# (2) Pooled OLS / Fixed Effects / Random Effects 추정 (수정본)
# =========================================

import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.panel import PooledOLS, PanelOLS, RandomEffects

# 종속변수/설명변수 지정
y = df["lwage"]

X_cols = ["educ", "exper", "expersq", "union", "married", "black", "hisp"]
X = df[X_cols]

# 상수항 추가 (Pooled/RE는 보통 상수 포함)
X_const = sm.add_constant(X)

# -------------------------
# 2-1) Pooled OLS
# -------------------------
pooled_mod = PooledOLS(y, X_const)
pooled_res = pooled_mod.fit(cov_type="clustered", cluster_entity=True)  # 개인 기준 군집표준오차

# -------------------------
# 2-2) Fixed Effects (개인 고정효과) - 방법 A (추천): 흡수 변수 자동 제거
# -------------------------
# drop_absorbed=True를 주면 educ/black/hisp 같은 시간불변 변수가 자동으로 빠지고 추정된다.
fe_mod = PanelOLS(y, X, entity_effects=True, drop_absorbed=True)
fe_res = fe_mod.fit(cov_type="clustered", cluster_entity=True)

# -------------------------
# 2-3) Random Effects
# -------------------------
re_mod = RandomEffects(y, X_const)
re_res = re_mod.fit(cov_type="clustered", cluster_entity=True)

# -------------------------
# 결과를 표 형태로 정리
# -------------------------
def tidy_table(res):
    out = pd.DataFrame({
        "coef": res.params,
        "std_err": res.std_errors,
        "t": res.tstats,
        "p": res.pvalues
    })
    return out

tbl_pooled = tidy_table(pooled_res)
tbl_fe = tidy_table(fe_res)
tbl_re = tidy_table(re_res)

print("\n# Pooled OLS\n")
print(tbl_pooled.to_markdown())

print("\n# Fixed Effects (drop_absorbed=True)\n")
print(tbl_fe.to_markdown())

print("\n# Random Effects\n")
print(tbl_re.to_markdown())

# FE에서 실제로 남은 변수(추정된 계수) 확인
print("\nFE에 남은 계수들:", list(fe_res.params.index))
```

```python
# =========================================
# (3) Hausman 검정 (FE vs RE)
# =========================================
# 개념:
# - H0: 개체효과와 설명변수가 독립 -> RE가 일치추정량(consisent)
# - H1: 독립 아님 -> RE는 불일치, FE를 선호
#
# 계산(표준적 형태):
#   diff = b_FE - b_RE
#   V = Var(b_FE) - Var(b_RE)
#   H = diff' * V^{-1} * diff  ~ chi2(k)
#
# 주의:
# - 군집(robust) 공분산을 쓰면 V가 비가역/비양정이 될 수 있음.
# - 그 경우(역행렬 실패)에는 비강건(non-robust) 공분산으로 다시 계산하는 방식을 자주 쓴다.
# =========================================

import numpy.linalg as la
from scipy import stats

def hausman_test(fe_res, re_res, common_only=True):
    # 1) 공통으로 존재하는 계수 이름 정리
    b_fe = fe_res.params
    b_re = re_res.params

    if common_only:
        common = b_fe.index.intersection(b_re.index)
    else:
        # common_only=False면 일치하지 않는 변수 때문에 보통 바로 깨진다.
        common = b_fe.index.intersection(b_re.index)

    # 2) 공통 계수만 추출
    b_fe_c = b_fe.loc[common]
    b_re_c = b_re.loc[common]

    # 3) 공분산행렬에서도 공통 부분만 추출
    V_fe = fe_res.cov.loc[common, common]
    V_re = re_res.cov.loc[common, common]

    # 4) Hausman 통계량 계산
    diff = (b_fe_c - b_re_c).to_numpy()
    V = (V_fe - V_re).to_numpy()

    # 역행렬(또는 의사역행렬) 사용
    try:
        V_inv = la.inv(V)
    except la.LinAlgError:
        V_inv = la.pinv(V)

    stat = float(diff.T @ V_inv @ diff)
    df_h = len(common)
    pval = 1 - stats.chi2.cdf(stat, df_h)

    return {
        "stat": stat,
        "df": df_h,
        "pvalue": pval,
        "common_params": list(common)
    }

# 1) 우선 군집표준오차 기반 결과로 Hausman 계산 시도
h = hausman_test(fe_res, re_res, common_only=True)

print("Hausman test (clustered cov) 결과")
print(h)

# 2) (선택) 만약 clustered cov로 결과가 불안정하다고 판단되면,
#    비강건 공분산(기본)으로 다시 적합해서 Hausman을 계산할 수도 있다.
#    아래는 그 예시.

fe_res_nr = fe_mod.fit()  # 기본(비강건) 공분산
re_res_nr = re_mod.fit()  # 기본(비강건) 공분산

h_nr = hausman_test(fe_res_nr, re_res_nr, common_only=True)

print("\nHausman test (non-robust cov) 결과")
print(h_nr)

# 3) 마크다운 표로 예쁘게 출력하고 싶으면:
hausman_tbl = pd.DataFrame(
    {"chi2": [h["stat"]], "df": [h["df"]], "pvalue": [h["pvalue"]]}
)
print("\n# Hausman (clustered)\n")
print(hausman_tbl.to_markdown(index=False))

hausman_tbl_nr = pd.DataFrame(
    {"chi2": [h_nr["stat"]], "df": [h_nr["df"]], "pvalue": [h_nr["pvalue"]]}
)
print("\n# Hausman (non-robust)\n")
print(hausman_tbl_nr.to_markdown(index=False))
```

## **Gemini 답변**

제공된 차트(시계열 데이터)와 호가창(Order Book 데이터) 이미지를 기반으로 주가 상승 확률을 실시간으로 추정하는 시스템을 구축하기 위한 방법론을 제안한다. 이 시스템은 **시장 미시구조(market microstructure)** 변수와 **기술적 지표(technical indicators)** 를 결합하여 단기 가격 변동을 예측하는 모델링이 핵심이다.

전체 프로세스는 **데이터 전처리 및 특성 공학(feature engineering)**, **모델링**, **시스템 아키텍처**의 3단계로 구성된다.

---

### 1. 데이터 정의 및 특성 공학 (Feature Engineering)

제공된 이미지는 크게 두 가지 정보 소스를 보여준다. 이를 수치화하여 모델의 입력 변수($X$)로 변환해야 한다.

#### A. 호가창(Order Book) 기반 변수 (이미지 2)
호가창 데이터는 초단기 가격 예측(수 초~수 분 내)에 가장 강력한 설명력을 가진다.

1.  **호가 불균형(Order Book Imbalance, OBI)**
    매수 잔량과 매도 잔량의 비율은 단기 수급을 나타내는 핵심 지표다.
    $$\text{OBI}_t = \frac{\sum V_{bid, t} - \sum V_{ask, t}}{\sum V_{bid, t} + \sum V_{ask, t}}$$
    * $V_{bid}$: 매수 잔량, $V_{ask}$: 매도 잔량
    * 제공된 이미지에서 매도 잔량이 254,388, 매수 잔량이 158,950이므로 현재 강한 매도 압력이 존재함을 수치화할 수 있다.

2.  **가중 중간 가격(Weighted Mid-Price, WMP)**
    단순 중간 가격보다 호가 잔량을 반영한 가격이 실제 체결 가격에 더 가깝다.
    $$P_{wmp} = \frac{P_{ask} \cdot V_{bid} + P_{bid} \cdot V_{ask}}{V_{bid} + V_{ask}}$$

3.  **호가 스프레드(Bid-Ask Spread)**
    $$S_t = P_{ask, t} - P_{bid, t}$$
    스프레드가 좁혀질 때 변동성이 폭발하는 경향이 있으므로 이를 변수로 사용한다.

4.  **가격 장벽(Price Walls) 탐지**
    이미지에서 161,500원에 49,372주의 대량 매도벽이 존재한다. 특정 호가에 $\ell_1$ 놈(norm) 기준으로 이상치(outlier) 수준의 물량이 쌓여있는지 여부를 이진 변수(0 또는 1) 혹은 비율 변수로 생성한다.

#### B. 차트 및 체결 기반 변수 (이미지 1)
과거의 가격 흐름을 통해 추세(trend)와 모멘텀(momentum)을 포착한다.

1.  **기술적 지표 (Technical Indicators)**
    * **RSI (Relative Strength Index):** 과매수/과매도 구간 식별.
    * **MACD:** 추세의 방향성 확인.
    * **VWAP (Volume Weighted Average Price):** 거래량 가중 평균 가격 대비 현재 가격의 위치.

2.  **거래량 강도 (Volume Intensity)**
    직전 $k$분 동안의 거래량 변화율. 이미지 하단의 거래량 스파이크는 정보의 유입을 암시하므로 중요한 변수가 된다.

3.  **변동성 (Volatility)**
    최근 $n$분간의 고가-저가 범위 또는 표준편차.

---

### 2. 확률 추정 모델링 (Modeling)

주가 상승 여부를 이진 분류(binary classification) 문제로 정의한다. 목표 변수 $Y$는 $t+k$ 시점의 가격이 현재보다 높은지($1$), 혹은 낮거나 같은지($0$)로 설정한다.

#### A. 모델 선택
1.  **Gradient Boosting Machine (XGBoost / LightGBM)**
    * **장점:** 정형 데이터(tabular data)에서 가장 우수한 성능을 보이며, 호가창 데이터의 비선형적 관계를 잘 학습한다. 과적합 방지를 위해 정규화(regularization) 파라미터를 튜닝해야 한다.
    * **특징:** 변수 중요도(feature importance)를 통해 어떤 호가 패턴이 상승에 기여했는지 해석 가능하다.

2.  **LSTM (Long Short-Term Memory) 또는 GRU**
    * **장점:** 시계열 데이터(차트의 흐름)를 순차적으로 학습하는 데 유리하다. 호가창의 변화 흐름(flow)을 잡아내는 데 효과적이다.

#### B. 학습 전략
* **라벨링(Labeling):** 단순히 $P_{t+1} > P_t$로 하면 노이즈가 심하다. "수수료를 제외하고 $x$% 이상 상승"과 같이 임계값을 설정하거나, 3-class 분류(상승, 하락, 횡보)로 접근하는 것이 안정적이다.
* **검증(Validation):** 시계열 데이터이므로 랜덤 셔플이 아닌 **Walk-forward Validation** (과거 데이터로 학습, 미래 데이터로 검증)을 사용해야 한다. 이때 붓스트랩(bootstrap) 기법을 활용하여 모델의 신뢰구간을 추정할 수 있다.

---

### 3. 실시간 시스템 아키텍처

실시간 추정을 위해서는 지연 시간(latency) 최소화가 필수적이다.

1.  **데이터 수집 (Ingestion):** 증권사 API의 WebSocket을 통해 `Real-time Tick` 데이터와 `Order Book` 변경 사항을 수신한다.
2.  **특성 계산 (Stream Processing):** 들어오는 데이터를 윈도우(window) 단위로 묶어 위에서 정의한 OBI, WMP 등의 지표를 실시간 계산한다.
3.  **추론 (Inference):** 사전 학습된 모델에 계산된 특성을 입력하여 상승 확률 $P(Y=1 | X)$를 출력한다.
4.  **실행 (Execution):** 확률이 특정 임계값(예: 0.7)을 넘으면 매수 신호를 발생시킨다.

---

### 4. 고려사항 및 주의점

* **방해 파라미터(Nuisance Parameter) 처리:** 시장 전체의 등락(KOSPI/KOSDAQ 지수 등)은 개별 종목 예측에 노이즈로 작용할 수 있다. 이를 조건부 제거(partialling out)하여 개별 종목의 순수한 움직임만 추출하는 기법을 고려할 수 있다.
* **오버피팅(Overfitting):** 금융 데이터는 노이즈가 매우 심하다. 모델 복잡도를 높이기보다 $\ell_1$ 또는 $\ell_2$ 페널티를 적용하여 일반화 성능을 높이는 데 주력해야 한다.
* **비정상성(Non-stationarity):** 시장의 분포는 시간에 따라 변한다(covariate shift). 주기적으로 모델을 재학습(retraining)하는 파이프라인이 필요하다.

## **ChatGPT 답변**

시스템 구축이 가능하다. 핵심은 “1분봉 + 호가창(레벨2) + 체결/거래량”을 실시간 피처로 바꾸고, 아주 짧은 미래 구간에 대해 라벨을 정의해서 $P(\text{up})$를 계속 업데이트하는 구조로 만들면 된다. 아래는 바로 구현 가능한 설계안이다.

---

### 1) 문제를 확률추정으로 정식화

시각 $t$에서 관측되는 상태를 $x_t$ (1분봉, 거래량, 호가/잔량, 스프레드 등)라고 하자. 목표는

$$
p_t = P(Y_t = 1 \mid x_t)
$$

여기서 $Y_t$는 “앞으로 $H$초 또는 $H$분 내 상승” 같은 사건이다.

라벨을 이렇게 잡는 게 실전에서 제일 안정적이다.

- 중간가격(mid)  
$$
m_t = \frac{\text{best\_ask}_t + \text{best\_bid}_t}{2}
$$

- 수평선(예측 구간) $H$ (예: 60초, 180초, 5분 등)  
- 임계치(미세 잡음 제거) $\delta$ (예: 1~2틱 또는 spread 기반)

라벨 예시:

$$
Y_t = 1\{ m_{t+H} - m_t \ge \delta \}
$$

3클래스(상승/하락/노트레이드)로 가면 더 현실적이다.

$$
Y_t \in \{+1, 0, -1\}
$$

- $+1$: $m_{t+H} - m_t \ge \delta$  
- $-1$: $m_{t+H} - m_t \le -\delta$  
- $0$: 그 사이(거의 횡보)

실시간 “상승 확률”은

$$
P(Y_t = +1 \mid x_t)
$$

로 출력하면 된다.

---

### 2) 입력 데이터에서 뽑아야 할 피처

이미지에 있는 2종류(1분 캔들, 전체 호가/잔량)를 그대로 숫자로 변환하면 된다. 추천 피처 묶음은 아래.

#### (A) 1분봉/거래량 기반

1분 단위로 업데이트되지만, “마지막 1분봉 + 직전 $k$개”로 시계열 피처를 만든다.

**수익률**
$$
r_t = \log\left(\frac{m_t}{m_{t-1}}\right)
$$

**캔들 구조**
- body = close − open  
- upper shadow = high − max(open, close)  
- lower shadow = min(open, close) − low  

**롤링 변동성**
$$
\sigma_t = \sqrt{\frac{1}{k}\sum_{i=0}^{k-1}(r_{t-i} - \bar{r})^2}
$$

추가:
- 거래량 변화율
- 거래대금 변화율
- VWAP 이탈
$$
\text{dev}_t = \frac{m_t - \text{VWAP}_t}{\text{tick}}
$$

---

#### (B) 호가/잔량 (레벨 $L$)

**스프레드**
$$
s_t = \text{ask1}_t - \text{bid1}_t
$$

**마이크로프라이스**
$$
\mu_t = \frac{\text{ask1}_t \cdot q^{bid1}_t + \text{bid1}_t \cdot q^{ask1}_t}{q^{bid1}_t + q^{ask1}_t}
$$

**1레벨 불균형**
$$
I^{(1)}_t = \frac{q^{bid1}_t - q^{ask1}_t}{q^{bid1}_t + q^{ask1}_t}
$$

**$L$레벨 누적 불균형**
$$
I^{(L)}_t = \frac{\sum_{i=1}^L q^{bidi}_t - \sum_{i=1}^L q^{aski}_t}{\sum_{i=1}^L q^{bidi}_t + \sum_{i=1}^L q^{aski}_t}
$$

추가:
- 오더북 기울기
- 가격대별 잔량 중심

---

#### (C) 변화율 (가장 중요)

호가 수준보다 **변화**가 핵심 신호다.

잔량 변화
$$
\Delta q^{bid1}_t = q^{bid1}_t - q^{bid1}_{t-\Delta}
$$

불균형 변화
$$
\Delta I^{(L)}_t = I^{(L)}_t - I^{(L)}_{t-\Delta}
$$

추가:
- 호가 갱신 빈도
- 1레벨 교체 횟수
- (체결 데이터 있으면) 매수/매도 체결 불균형

---

### 4) 실시간 시스템 구조

### 5) 성능평가

로그손실
$$
\text{LogLoss} = -\frac{1}{n}\sum_{i=1}^n \left(y_i\log p_i + (1-y_i)\log(1-p_i)\right)
$$

Brier score
$$
\text{Brier} = \frac{1}{n}\sum_{i=1}^n (p_i - y_i)^2
$$

캘리브레이션:
- 예측확률 0.7 → 실제 70% 상승 여부 확인

백테스트:
- 스프레드
- 수수료
- 슬리피지
- 체결 실패 포함

## 2. 질문

**숫자 데이터 대신 시각적 패턴을 학습시키는 방식도 가능?**

</div>

## **Gemini 답변**

네, 가능하다. 이를 **"Computer Vision based Algorithmic Trading"** 또는 **"Visual Pattern Recognition"** 이라고 한다.

숫자 데이터를 직접 넣는 것이 아니라, 사람이 차트를 보고 판단하는 것처럼 **이미지(픽셀) 자체를 인공지능(CNN 등)에 입력하여 시각적 패턴을 학습시키는 방식**이다.

크게 세 가지 접근법이 있다.

---

### 1. 차트 이미지 분석 (2D CNN 활용)
가장 직관적인 방법으로, 제공해주신 첫 번째 이미지(캔들 차트)를 그대로 모델의 입력으로 사용하는 것이다.

* **핵심 원리:**
    이미지 인식 분야에서 쓰이는 **CNN(Convolutional Neural Networks)** 모델(ResNet, EfficientNet 등)을 사용한다. 모델은 픽셀 덩어리에서 '상승 깃발형', '헤드 앤 숄더', '이중 바닥' 같은 기하학적 패턴(feature map)을 스스로 학습한다.
* **프로세스:**
    1.  과거 10년 치 차트를 일정한 크기(예: $224 \times 224$ 픽셀)의 이미지 파일로 저장한다.
    2.  각 이미지가 발생한 직후 주가가 상승했는지 하락했는지 라벨링($Y$)한다.
    3.  CNN 모델에 이미지를 입력해 $P(Y=\text{상승} | \text{이미지})$를 학습시킨다.
* **장점:** 수치 데이터로는 표현하기 힘든 복합적인 패턴(추세선의 각도, 지지/저항선의 시각적 형태 등)을 포착할 수 있다.

---

### 2. 호가창 시각화 (Limit Order Book to Image)
두 번째 이미지인 호가창은 숫자가 많아 OCR(광학 문자 인식)로 읽기보다, **히트맵(Heatmap)** 형태로 변환하여 분석하는 것이 일반적이다.

* **핵심 원리:**
    호가창의 '가격(행)'과 '시간(열)'을 축으로 하고, '잔량(Volume)'을 '색상의 진하기(Pixel Intensity)'로 표현하여 하나의 이미지로 만든다.
* **분석 방법:**
    * 시간이 지남에 따라 호가 잔량이 쌓이고 사라지는 것을 이미지상의 **'흐름(Flow)'**이나 **'벽(Wall)'** 형태로 변환한다.
    * 예를 들어, 매도 잔량이 갑자기 진한 색(대량 주문)으로 나타났다가 사라지는 패턴(허수 주문, Spoofing)을 CNN이 감지해낸다.
* **연구 사례:** *Jiang et al. (2017)* 등의 연구에서는 호가창 데이터를 이미지화하여 CNN에 넣었을 때, 기존 수치 모델보다 단기 가격 예측력이 높다는 결과를 제시했다.

---

### 3. 시계열의 이미지 변환 (GAF / MTF)
차트 스크린샷을 찍는 것이 아니라, 시계열 데이터(가격 숫자)를 수학적으로 변환하여 이미지를 생성하는 고도화된 방법이다.

* **GAF (Gramian Angular Field):** 시간 순서대로 나열된 가격 데이터를 극좌표계로 변환하여 행렬(이미지)로 만든다. 시간적 상관관계를 시각적 패턴으로 보존한다.
* **MTF (Markov Transition Field):** 가격 변동의 전이 확률을 이미지 픽셀 값으로 매핑한다.
* 이 방식은 노이즈가 많은 스크린샷(축, 글자 등 불필요한 정보) 대신 순수한 데이터의 구조적 이미지를 얻을 수 있어 학술적으로 가장 많이 쓰인다.

---

### 픽셀 기반 방식의 장단점

| 구분 | 장점 (Pros) | 단점 (Cons) |
| :--- | :--- | :--- |
| **직관성** | 사람이 차트를 보는 방식과 유사하게 '형태'를 인지함. | 161,000원과 161,100원의 미세한 차이를 픽셀로는 구별하기 어려움 (해상도 문제). |
| **패턴 포착** | 비선형적이고 복잡한 패턴(수급 공백 등)을 잘 찾음. | 모델이 무거워 실시간 추론(Inference) 속도가 느릴 수 있음. |
| **전처리** | 복잡한 금융 공학적 변수 생성(Feature Engineering)이 덜 필요함. | 학습 데이터(이미지) 생성 및 저장에 막대한 용량이 필요함. |

---

### 결론 및 제언

주신 이미지 데이터를 이용해 **"이미지 기반 시스템"**을 구축한다면 다음 단계가 필요하다.

1.  **전처리:** 스크린샷에서 불필요한 UI(버튼, 메뉴 등)를 잘라내고(Crop), 차트 영역과 호가 영역만 남긴다.
2.  **모델링:**
    * 차트 이미지 $\rightarrow$ **CNN (VGG16 or ResNet)**
    * 호가창 이미지 $\rightarrow$ **3D CNN** (시간 흐름을 포함한 비디오 형식으로 처리)
3.  **앙상블:** 이미지에서 추출한 벡터와, 수치 데이터(가격, 거래량)를 결합(Concatenate)하여 최종 예측을 하는 것이 단일 모델보다 성능이 좋다.

## **ChatGPT 답변**

가능하다. 숫자 피처 대신 **차트·호가를 이미지로 변환하여 시각적 패턴을 직접 학습**시키는 방식은 실제로 사용되며, 특히 단기 트레이딩에서는 다음 세 가지 접근이 대표적이다.

다만 결론부터 말하면,

- 가능함 (기술적으로 문제 없음)
- 그러나 **숫자 기반 모델보다 항상 좋은 것은 아님**
- 효과는 **패턴의 구조성과 데이터량**에 크게 의존한다

아래에 구조적으로 설명한다.

---

## 1) 문제 정식화 (이미지 기반)

시각 $t$에서 상태를 숫자 벡터 $x_t$ 대신 이미지 $I_t$로 정의한다.

목표:

$$
p_t = P(Y_t = +1 \mid I_t)
$$

라벨 정의는 동일하다.

예:
$$
Y_t = 1\{ m_{t+H} - m_t \ge \delta \}
$$

즉,

- 입력: 차트 이미지 또는 호가 시각화
- 출력: 향후 상승 확률

---

## 2) 어떤 시각 데이터를 사용할 수 있는가

### (A) 캔들 차트 이미지

예: 최근 30~60개의 1분봉을 이미지로 생성

구성:
- OHLC 캔들
- 거래량 히스토그램
- 이동평균선
- VWAP

입력 형태:
- 크기: 예) 128×128 또는 224×224
- 채널: RGB 또는 grayscale

학습 모델:
- CNN (ResNet, EfficientNet 등)

이 방식은 “헤드앤숄더”, “돌파”, “추세 전환” 같은 **형태 패턴**을 자동 학습한다.

---

### (B) 오더북 히트맵 (강력 추천)

호가 데이터를 시간–가격 공간에 시각화한다.

구성:
- x축: 시간
- y축: 가격 레벨
- 색상: 잔량 크기

또는:

- bid: 파란색
- ask: 빨간색

이것은 사실상 **시장 유동성의 영상 데이터**이며, 단기 방향 예측에 매우 유용하다.

---

### (C) 통합 멀티패널 이미지

한 장에 함께 표현:

- 상단: 가격 캔들
- 중단: 거래량
- 하단: 오더북 히트맵

즉,

$$
I_t = \text{Chart} + \text{Volume} + \text{Orderbook}
$$

---

## 3) 모델 구조

### (1) CNN 기반 분류

가장 기본적인 구조:

$$
I_t \rightarrow \text{CNN} \rightarrow p_t
$$

예:
- ResNet18
- EfficientNet-B0
- MobileNet (실시간용)

손실함수:
- Binary cross entropy
- 또는 focal loss (클래스 불균형 대응)

---

### (2) CNN + 시계열 결합

이미지 여러 장을 시퀀스로 사용:

$$
(I_{t-k}, \dots, I_t)
$$

구조:

- CNN → feature
- LSTM 또는 Transformer → 시간 패턴
- Output → 상승 확률

---

## 4) 장점과 단점

### 장점

1. **피처 엔지니어링 불필요**
   - 모델이 패턴을 자동 학습

2. **비선형 구조 포착**
   - 추세 채널
   - 돌파 구조
   - 유동성 벽

3. **오더북 구조 표현에 강함**

---

### 단점 (매우 중요)

#### (1) 데이터 요구량이 큼

CNN은 최소:

- 수십만 ~ 수백만 샘플 필요

---

#### (2) 정보 손실 가능

이미지 변환 과정에서:

- 정확한 가격
- 미세한 변화
- 잔량 비율

이 정보가 왜곡될 수 있다.

숫자 모델은

$$
I^{(L)}_t = \frac{Q^{bid} - Q^{ask}}{Q^{bid} + Q^{ask}}
$$

같은 정밀 정보를 그대로 사용한다.

---

#### (3) 실시간 비용

- 이미지 생성 비용
- GPU 필요
- 지연(latency) 증가

초단타에서는 치명적일 수 있다.

---

## 5) 실전에서 가장 효과적인 방식

완전 이미지 모델보다 **하이브리드**가 성능이 좋다.

구조:

$$
\text{CNN(image)} \rightarrow z_t
$$

$$
x_t = [z_t, \text{numeric features}]
$$

$$
x_t \rightarrow LightGBM \text{ 또는 MLP}
$$

즉,

- CNN: 패턴 인식
- 숫자: 미세구조 정보
- 최종 모델: 확률 출력

---

## 6) 언제 이미지 방식이 유리한가

다음 조건이면 고려할 가치가 크다.

- 오더북 히트맵 사용 가능
- 초단타 (수초~수분)
- 데이터 대량 확보 가능
- GPU 사용 가능

반대로,

- 데이터가 적거나
- 실행 지연이 중요하거나
- 운영 단순성이 중요하면

숫자 기반 LightGBM이 더 안정적이다.

---

## 7) 결론

이미지 기반 학습은 충분히 가능하며 다음과 같이 정식화된다.

$$
p_t = P(Y_t = +1 \mid I_t)
$$

그러나 실전 성능 기준으로 보면,

- 단독 CNN < 숫자 기반 모델
- **CNN + 숫자 결합 모델**이 가장 강력

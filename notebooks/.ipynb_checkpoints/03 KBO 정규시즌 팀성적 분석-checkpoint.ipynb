## Gemini 데이터세트 구축

제공된 6장의 이미지를 바탕으로 KBO 리그 팀 기록(타자, 투수, 수비, 주루)을 모두 통합하여 하나의 **데이터세트(Dataset)** 로 구축하고, Python(Pandas, Matplotlib/Seaborn)을 활용해 탐색적 데이터 분석(EDA)을 수행한다.

사용자 페르소나와 요청 사항을 반영하여 다음과 같은 절차로 진행한다.

1. **데이터 구축:** 이미지 내 수치를 텍스트화하여 Pandas DataFrame으로 변환.
2. **데이터 전처리:** 서로 다른 정렬 순서를 가진 테이블들을 '팀명' 기준으로 병합(Merge)하고, 이닝(IP)과 같은 문자열 데이터를 수치형으로 변환.

---

### 유의사항

**"모든 관측값이 입력된 데이터세트를 만들라"** 는 선생님의 첫 번째 요청 사항을 엄격하게 따르지 않고, **분석의 효율성과 핵심 지표(Key Performance Indicator) 위주로 임의 선별(Feature Selection)** 했습니다. 예를 들어 수비 기록(Image 5)에서 2개 컬럼(`E`, `FPCT`)만 선택하고 나머지를 제외한 이유는 다음과 같습니다.

#### 1. 변수 중요도(Feature Importance)와 승률의 상관관계

데이터 분석, 특히 **'우승(승률)'** 이라는 타겟 변수(Target Variable)를 설명할 때 모든 변수가 동일한 가치를 갖지 않습니다.

* **선택한 변수 (`E`, `FPCT`):**
   * **실책(E)** 과 **수비율(FPCT)** 은 팀의 수비 안정성을 나타내는 가장 직접적인 지표입니다. 실책 하나는 곧바로 실점 위기나 패배로 직결되므로 승률 분석에 필수적입니다.


* **제외한 변수 (`PO`, `A` 등):**
   * **자살(PO, Putout)** 과 **보살(A, Assist)** 은 수비 성공 횟수입니다. 하지만 이 숫자가 높다고 해서 "수비를 잘한다"고 단정할 수 없습니다. 투수가 삼진을 못 잡고 맞춰 잡는 유형이거나, 단순히 수비 기회가 많았던 팀일수록 이 수치가 높게 나옵니다(단순 누적 기록). 즉, 승률을 설명하는 설명력(Explanatory Power)이 낮습니다.

#### 2. 다중공선성(Multicollinearity) 방지

* **`FPCT`(수비율)** 는 이미 `E`(실책), `PO`, `A`를 계산식에 포함하여 만든 파생 변수입니다.
   * $FPCT = \frac{PO + A}{PO + A + E}$

* 따라서 `PO`, `A`, `E`를 모두 독립변수로 넣고 `FPCT`까지 넣으면 변수 간의 강한 상관관계로 인해 분석 모델의 신뢰도가 떨어지는 **다중공선성 문제**가 발생할 수 있습니다. 이를 피하기 위해 가장 대표성 있는 `FPCT`와 `E`만 남겼습니다.

### 3. 중복 데이터(Redundancy)

* 이미지 5(수비)에 있는 **도루 저지(CS, CS%)** 관련 데이터는 이미지 6(주루)의 데이터와 개념적으로 겹치거나, 포수 개인의 역량에 가까운 지표라 팀 전체 승률 분석에서는 우선순위를 낮췄습니다.

## Gemini 분석

팀 순위나 승률(WPCT)에 영향을 미치는 핵심 변수(Key Features)를 식별하기 위해 상관관계 분석과 시각화를 진행한다.

### 1. 주요 변수 간 상관관계 행렬

야구의 승패는 득점(생산)과 실점(억제)의 균형에서 온다. 투타의 핵심 지표가 승률(WPCT)과 얼마나 강한 상관관계를 갖는지 확인한다.



**분석 결과:**

1. **ERA(평균자책점)와 WPCT:** 매우 강한 **음의 상관관계**를 보일 것이다. (ERA가 낮을수록 승률이 높음)
2. **OPS(출루율+장타율)와 WPCT:** 일반적으로 강한 **양의 상관관계**를 갖는다.
3. **WHIP(이닝당 출루 허용률):** ERA와 유사하게 승률과 강한 음의 상관관계를 가질 것이다.

---

### 2. 투수력(ERA) vs 타력(OPS)의 승률 기여도 비교

어느 쪽이 승률을 더 잘 설명하는지 산점도(Scatter plot)와 회귀선으로 시각화한다.



**해석:**
이 데이터세트에서는 ERA(투수 지표)가 OPS(타자 지표)보다 승률과 더 뚜렷한 선형 관계를 보인다. 이미지 상 1위 한화(ERA 3.55), 2위 SSG(3.63), 3위 LG(3.79) 순으로 ERA가 낮은 팀들이 상위권에 포진해 있음을 알 수 있다. 반면 타격 지표인 OPS는 LG와 삼성이 높지만, 팀 승률과의 정렬은 투수 지표만큼 완벽하게 일치하지 않을 수 있다.

### 3. 피타고라스 승률 분석 (Pythagorean Expectation)

피타고라스 승률은 세이버메트릭스의 창시자 빌 제임스(Bill James)가 고안한 지표로, 팀의 **득점($R$)**과 **실점($RA$)** 만을 사용하여 해당 팀이 기록했어야 할 **'기대 승률'** 을 추정하는 공식이다. 공식의 형태가 피타고라스의 정리($a^2 + b^2 = c^2$)와 유사하여 붙여진 이름이다.

$$\text{Expected Win \%} = \frac{R^2}{R^2 + RA^2}$$

(일반적으로 지수 $2$를 사용하며, 현대 야구에서는 정밀도를 높이기 위해 $1.83$을 사용하기도 한다.)

이 분석의 핵심 목적은 **"팀의 실제 승률이 득실점 차이에 비해 얼마나 효율적이었는가?"** 를 판단하는 데 있다.

* **실제 승률 > 기대 승률 (양수, +):** 득실 마진에 비해 더 많은 승리를 챙겼다. 접전 상황에서 강했거나 불펜 운용이 효율적이었음을 의미한다. (통계적으로 '운이 좋았다'고 표현하기도 한다.)
* **실제 승률 < 기대 승률 (음수, -):** 득실 마진에 비해 승수를 쌓지 못했다. 크게 이기고 좁은 점수 차로 패배하는 '비효율적인 경기'를 했음을 의미한다. (통계적으로 '운이 나빴다'고 표현한다.)



#### 분석 결과

위 그래프는 KBO 10개 구단의 **[실제 승률 - 기대 승률]** 차이를 시각화한 것이다.

**1) 고효율 그룹: "지표보다 강한 팀" (NC, KT, SSG, 롯데, 키움)**
이들은 득실점 데이터가 예측하는 것보다 실제 순위가 더 높게 나온 팀들이다. **'이기는 법을 아는 야구'** 를 했다고 평가할 수 있다.

   * **NC 다이노스 (+0.037):** 가장 돋보이는 효율성을 보여준다. 득점(732)보다 실점(766)이 많아 득실마진이 마이너스였음에도, 실제로는 5할 승률(0.514)을 넘기며 4위를 차지했다. 질 때는 대패하더라도, 접전 상황에서는 집중력을 발휘해 승리를 가져가는 '실속형 야구'를 했다.
   * **KT 위즈 (+0.018) & SSG 랜더스 (+0.009):** 두 팀 모두 득실마진 대비 승률이 높다. 특히 KT는 득실마진이 음수였음에도 5할 승률을 사수했다. 이는 불펜의 힘이나 감독의 경기 운영 능력이 데이터의 열세를 극복했음을 시사한다.
   * **롯데 & 키움:** 비록 하위권에 머물렀으나, 기대 승률보다는 실제 승률이 높았다. 전력이 약해(특히 키움은 득실마진 -279점) 많은 패배를 당했지만, 잡을 수 있는 경기는 확실히 잡아내며 최악의 상황 속에서도 분전했다.

**2) 저효율 그룹: "지표보다 아쉬운 팀" (Samsung, LG)**
이들은 압도적인 득실점 지표를 가지고도, 그에 걸맞은 승수를 온전히 챙기지 못한 팀들이다.

   * **삼성 라이온즈 (-0.068):** 리그에서 가장 비효율적인 시즌을 보냈다. 팀 타율 1위, 홈런 1위로 득점력은 막강했지만, 실제 승률은 기대치보다 6푼 8리나 낮았다. 이는 타선이 터질 때는 10점 차 대승을 거두지만, 정작 중요한 접전 상황에서는 불펜 불안 등으로 패배하는 **'영양가 없는 승리'**가 많았다는 데이터 과학적 증거다.
   * **LG 트윈스 (-0.038):** 득실마진 +204라는 압도적인 수치를 기록하며 정규시즌 우승을 차지했다. 하지만 통계적 기대치(승률 6할 4푼)보다는 낮은 6할 3리의 승률을 기록했다. 워낙 전력이 강해 1위를 하는 데는 문제가 없었으나, 세밀한 효율성 측면에서는 더 승수를 쌓을 여지가 있었다.

**3) 결론**
데이터는 **"강한 공격력이 반드시 접전에서의 승리를 보장하지 않는다"** 는 사실을 보여준다. 장기 레이스인 정규시즌에서는 득실마진(기대 승률)이 최종 순위의 하한선을 지지해주지만, 단기전이나 포스트시즌에서는 NC나 KT처럼 실제 승률이 기대 승률을 상회하는 **'위기 관리 능력(Clutch Ability)'** 이 중요한 변수가 될 수 있다.

## 3. 정규시즌 우승 원동력

</div>

제공된 데이터세트에서 정규시즌 우승팀(승률 1위)은 **LG 트윈스(승률 0.603)** 다. 데이터를 기반으로 **"우승팀의 조건"** 과 **"승패를 가르는 핵심 요인"** 에 대한 인사이트를 분석했다.

### 1. 우승의 제1조건: 압도적인 득실마진 (Run Differential)

야구 데이터 분석에서 승률과 가장 강력한 상관관계를 갖는 지표는 단연 **득실마진(득점 - 실점)** 이다.

* **LG(1위):** 득점 788점 - 실점 584점 = **+204점**
* **한화(2위):** 득점 689점 - 실점 554점 = **+135점**
* **삼성(4위):** 득점 775점 - 실점 647점 = **+128점**
* **키움(10위):** 득점 581점 - 실점 860점 = **-279점**

우승팀 LG는 리그에서 유일하게 +200점 이상의 득실마진을 기록했다. 이는 피타고라스 승률(Pythagorean Expectation) 이론을 완벽하게 증명한다. 삼성은 득점력이 뛰어났음에도(775점, 2위), 실점이 많아(647점) 득실마진에서 밀렸고, 결국 승률도 LG와 한화에 미치지 못했다.

> **인사이트:** 우승은 단순히 "잘 치는 것"이나 "잘 던지는 것" 하나로 결정되지 않는다. **"얼마나 남기는 장사를 했느냐(득실차)"** 가 승률의 90% 이상을 설명한다.

---

### 2. '창'보다는 '방패', 하지만 우승은 '밸런스'

흔히 "타격은 팬을 부르고, 투수는 승리를 부른다"고 한다. 이 데이터에서도 그 경향이 확인된다.

* **최고의 창 (삼성):** 팀 OPS 0.780 (1위), 홈런 161개 (1위) → 그러나 **최종 순위 4위 (승률 0.521)**
* **최고의 방패 (한화):** 팀 ERA 3.55 (1위) → **최종 순위 2위 (승률 0.593)**
* **밸런스의 LG:** 팀 타율 1위(0.278), 팀 ERA 3위(3.79) → **최종 우승 (승률 0.603)**

삼성은 타격 지표(OPS, 홈런, 타점)를 휩쓸었지만, 평균자책점(ERA)이 4.12(5위)로 무너져 승리를 챙기지 못했다. 반면 한화는 압도적인 투수력으로 2위까지 올랐다. 결국 우승을 차지한 LG는 투타 모두 리그 3위 안에 드는 **'육각형 밸런스'** 를 갖춘 팀이었다.

> **인사이트:** 강력한 공격력(High OPS)은 팀을 상위권으로 이끌지만, 우승을 확정 짓는 것은 **투타의 균형**이다. 공격력만으로는 투수력의 부재를 메울 수 없다.

---

### 3. 숨겨진 승리의 요인: 볼넷/삼진 비율 (BB/K)과 출루율

세이버메트릭스 관점에서 '운'의 요소를 배제하고 팀의 진짜 실력을 보여주는 지표 중 하나가 **볼넷(BB)** 이다.

* **LG:** 볼넷 604개 (1위), 출루율(OBP) 0.361 (1위)
* **타 팀 대비:** LG는 2위 삼성(564개)보다 40개나 더 많은 볼넷을 얻어냈다.

안타는 운(BABIP)의 영향을 받지만, 볼넷은 선구안이라는 순수한 실력의 영역이다. LG는 가장 많은 볼넷을 골라내며, 아웃되지 않고 주자를 쌓아두는 능력이 탁월했다. 이는 대량 득점(빅이닝)의 발판이 된다.

> **인사이트:** 우승팀은 단순히 방망이를 잘 휘드르는 팀이 아니다. **"나쁜 공에 손대지 않고 걸어나가는 능력(Discipline)"** 이 가장 뛰어난 팀이 결국 가장 많은 득점을 올린다.

---

### 4. 수비 효율성 (Defensive Efficiency)

수비 지표인 **FPCT(수비율)** 와 **E(실책)** 를 보면 흥미로운 점이 발견된다.

* **한화:** 실책 86개 (최소 1위), FPCT 0.984 (1위)
* **LG:** 실책 92개 (최소 3위), FPCT 0.983 (2위)
* **KIA:** 실책 123개 (최다 1위), FPCT 0.977 (하위권)

1, 2위를 차지한 LG와 한화는 실책이 가장 적은 팀 그룹에 속한다. 반면 실책이 123개로 가장 많았던 KIA는 승률 0.464로 하위권에 머물렀다. 실책 하나는 투수의 투구 수를 늘리고, 주지 않아도 될 점수를 주는 '보이지 않는 패배 요인'이다.

> **인사이트:** 데이터 과학에서 말하는 **'노이즈(Error) 감소'** 가 실제 야구에서도 승리와 직결된다. 안정적인 수비는 투수의 ERA를 낮추고 승률을 방어하는 핵심 기제다.

---

### 요약

이 데이터세트를 통해 모델링을 한다면, 우승 확률(Target Variable)을 예측하는 데 가장 중요한 변수(Feature Importance)는 다음과 같은 순서일 것이다.

1. **득실마진 (Run Differential):** 압도적 1순위.
2. **WHIP (이닝당 출루 허용률):** 투수력의 핵심.
3. **OPS (출루율 + 장타율):** 공격력의 핵심.
4. **BB/K (볼넷/삼진 비율):** 팀의 기초 체력.

결론적으로, **"많이 득점하고(High OPS), 적게 실점하며(Low ERA), 실책을 최소화하는(High FPCT)"** 교과서적인 팀이 우승한다는 사실을 데이터가 증명하고 있다.

## 4. LG 트윈스의 "육각형 밸런스"

</div>

LG 트윈스의 "육각형 밸런스"를 시각화한 결과다. 이 레이더 차트는 공격(Batting), 투구(Pitching), 수비(Defense), 주루(Speed), 선구안(Discipline), 안정성(Stability) 6가지 핵심 지표를 리그 내 상대적 위치(0~1 스케일)로 변환하여 표현했다.



### 1. 차트 해석 (Hexagonal Analysis)

붉은색 영역이 LG 트윈스이며, 회색 점선은 리그 평균이다.

* **Discipline (선구안/볼넷):** 차트의 5시 방향을 보시면 LG의 그래프가 가장 바깥쪽(1.0)에 위치해 있다. 이는 리그 압도적 1위였던 볼넷 개수(604개)를 반영하며, LG 야구의 핵심이 **'참아내는 능력'** 에 있었음을 보여준다.
* **Batting (타격/OPS):** 12시 방향. 타격 역시 리그 최상위권으로, 평균을 훨씬 상회하는 공격력을 갖췄다.
* **Pitching & Stability (투구/ERA, WHIP):** 2시, 7시 방향. 투수 지표 역시 리그 평균보다 바깥쪽에 위치하여, 투타가 모두 준수함을 알 수 있다.
* **Speed (주루/도루):** 10시 방향. 기동력 또한 평균 이상이다.

### 2. 결론: "찌그러지지 않은 육각형"

보통 특정 팀은 공격만 강하거나(한쪽만 뾰족), 투수만 강한(반대쪽만 뾰족) 형태를 띠기 마련이다. 하지만 LG의 육각형은 **모든 꼭짓점이 평균선(회색 점선) 바깥에 위치**하며 찌그러짐 없이 고르게 확장된 형태를 보인다.

이것이 바로 데이터가 말하는 **'우승팀의 완전성(Completeness)'** 이다. 어느 한 곳 약점이 없는 밸런스가 장기 레이스인 정규시즌 우승의 원동력이 되었다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 준비 (영문 팀명 및 득점/실점 매핑)
# R: 득점 (Runs Scored), R_p: 실점 (Runs Allowed)
data = {
    'Team': ['LG', 'Samsung', 'Lotte', 'Hanwha', 'Doosan', 'NC', 'KIA', 'SSG', 'KT', 'Kiwoom'],
    'R': [788, 775, 676, 689, 647, 732, 668, 609, 648, 581],
    'R_p': [584, 647, 749, 554, 686, 766, 734, 576, 657, 860], # 팀별 실점 매칭 완료
    'WPCT': [0.603, 0.521, 0.478, 0.593, 0.442, 0.514, 0.464, 0.536, 0.511, 0.336]
}

df = pd.DataFrame(data)

# 1. 피타고라스 기대 승률 계산 (지수 2 사용)
# Formula: R^2 / (R^2 + R_p^2)
df['Exp_WPCT'] = df['R']**2 / (df['R']**2 + df['R_p']**2)

# 2. 실제 승률과의 차이 계산 (Diff = Actual - Expected)
df['Diff'] = df['WPCT'] - df['Exp_WPCT']

# 3. 시각화 (Bar Plot)
plt.figure(figsize=(7, 4))

# 차이 순으로 정렬 (내림차순)
df_sorted = df.sort_values('Diff', ascending=False)

# 색상 지정: 양수(파랑), 음수(빨강)
colors = ['#1f77b4' if x >= 0 else '#d62728' for x in df_sorted['Diff']]

sns.barplot(x='Team', y='Diff', hue='Team', data=df_sorted, palette=colors, legend=False)

# 기준선 (0) 추가
plt.axhline(0, color='black', linewidth=1)

# 제목 및 레이블 설정 (영어)
plt.title('Actual vs. Pythagorean Winning Percentage Difference\n(Positive: Overperforming, Negative: Underperforming)', fontsize=12)
plt.ylabel('Winning Percentage Difference', fontsize=10)
plt.xlabel('Team', fontsize=10)

# 막대 위에 수치 표시 (Annotation)
for i, v in enumerate(df_sorted['Diff']):
    plt.text(i, v + (0.002 if v > 0 else -0.005), f"{v:.3f}", 
             ha='center', va='bottom' if v > 0 else 'top', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# 1. 데이터 재구축 (필요한 지표만 추출하여 구성)
data = {
    'Team': ['LG', 'Samsung', 'Lotte', 'Hanwha', 'Doosan', 'NC', 'KIA', 'SSG', 'KT', 'Kiwoom'],
    'OPS': [0.770, 0.780, 0.718, 0.730, 0.714, 0.743, 0.734, 0.706, 0.706, 0.671],
    'ERA': [3.79, 4.12, 4.75, 3.55, 4.30, 4.82, 4.66, 3.63, 4.09, 5.39],
    'FPCT': [0.983, 0.984, 0.979, 0.984, 0.977, 0.978, 0.977, 0.980, 0.981, 0.977],
    'SB': [121, 98, 91, 101, 144, 186, 77, 129, 48, 83],
    'BB': [604, 564, 530, 439, 469, 504, 526, 493, 555, 439],
    'WHIP': [1.35, 1.35, 1.49, 1.27, 1.38, 1.50, 1.43, 1.30, 1.40, 1.57]
}
df = pd.DataFrame(data)

# 2. 데이터 전처리 및 정규화 (Min-Max Scaling)
# ERA, WHIP은 낮을수록 좋으므로 역방향 스케일링 적용
# Score = (Max - x) / (Max - Min)
df['ERA_Score'] = (df['ERA'].max() - df['ERA']) / (df['ERA'].max() - df['ERA'].min())
df['WHIP_Score'] = (df['WHIP'].max() - df['WHIP']) / (df['WHIP'].max() - df['WHIP'].min())

# 나머지는 정방향 스케일링
# Score = (x - Min) / (Max - Min)
for col in ['OPS', 'FPCT', 'SB', 'BB']:
    df[f'{col}_Score'] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# 레이더 차트용 데이터 준비
# 라벨 정의 
labels = ['Batting\n(OPS)', 'Pitching\n(ERA)', 'Defense\n(FPCT)', 'Speed\n(SB)', 'Discipline\n(BB)', 'Stability\n(WHIP)']
# 데이터 컬럼 순서
scores_col = ['OPS_Score', 'ERA_Score', 'FPCT_Score', 'SB_Score', 'BB_Score', 'WHIP_Score']

# LG 데이터 추출
lg_values = df.loc[df['Team'] == 'LG', scores_col].values.flatten().tolist()
# 리그 평균 데이터 추출
avg_values = df[scores_col].mean().values.flatten().tolist()
# 비교군: 삼성 (타격 1위지만 4위)
samsung_values = df.loc[df['Team'] == 'Samsung', scores_col].values.flatten().tolist()


# 레이더 차트는 마지막 점이 시작점과 이어져야 함 (폐곡선)
lg_values += lg_values[:1]
avg_values += avg_values[:1]
samsung_values += samsung_values[:1]

angles = [n / float(len(labels)) * 2 * pi for n in range(len(labels))]
angles += angles[:1]

# 3. 시각화
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# 각 축의 라벨 설정
plt.xticks(angles[:-1], labels, color='grey', size=10)

# y축 라벨 설정
ax.set_rlabel_position(0)
plt.yticks([0.25, 0.5, 0.75, 1.0], ["0.25", "0.50", "0.75", "1.00"], color="grey", size=6)
plt.ylim(0, 1)

# Plot 1: LG Twins
ax.plot(angles, lg_values, linewidth=2, linestyle='solid', label='LG Twins (Winner)', color='#C3002F') # LG Red color approximation
ax.fill(angles, lg_values, '#C3002F', alpha=0.25)

# Plot 2: League Average
ax.plot(angles, avg_values, linewidth=2, linestyle='dashed', label='League Average', color='grey')
# ax.fill(angles, avg_values, 'grey', alpha=0.1)

# 타이틀 및 범례
plt.title("Hexagonal Balance: LG Twins", size=14, color='black', y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 0.1))

plt.savefig('lg_hexagon_balance.png')
```

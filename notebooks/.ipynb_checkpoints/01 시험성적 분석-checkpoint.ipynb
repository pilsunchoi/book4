## ChatGPT 분석

요약하면, **추가신청(=1) 수강생들의 평균 총점이 더 낮게 나오긴 했지만**, 이 데이터만으로 **“추가신청이라서 성적이 낮다”** 고 말할 만큼 차이가 뚜렷하진 않다.

### 1) 기술통계 비교(총점 = (중간+기말)/2)

| 구분     | 표본수 | 총점 평균 | 총점 표준편차 | 총점 중앙값 |

* 따라서 “추가신청 수강생이 상대적으로 낮다”는 말은 **기술통계 수준에서는 맞지만**, **일관된 차이라고 결론내리긴 어렵다**.

## 3. 추가신청 수강생들의 성적이 상대적으로 낮은가? (계속)

**다음에 대해 좀 더 알기 쉽게 설명해줘.**

* Welch t-검정 기준(분산 다를 수 있음):

  * t ≈ **-0.71**, p ≈ **0.49**
  * 평균 차이 95% CI: **[-14.27, 7.01]**
* 효과크기(Cohen’s d) ≈ **-0.21** (작은 차이)

</div>

## ChatGPT 설명

좋다. 이건 “차이가 있어 보이는데, 그게 진짜 차이냐 아니면 우연이냐?”를 판단하는 부분이다. 하나씩 풀어보자.

---

### 1️⃣ Welch t-검정이 뭘 하는가

우리가 비교한 것:

> 추가신청 0 집단 평균 = 73.13
> 추가신청 1 집단 평균 = 69.50

겉으로 보면 **추가신청 집단이 3.63점 낮다**.

Welch t-검정은 질문을 이렇게 바꾼다:

> 이 차이가 **진짜 집단 차이** 때문인가,
> 아니면 **우연한 표본 변동** 때문인가?

Welch를 쓰는 이유는 두 집단의 분산(점수 흩어짐)이 같다고 가정하지 않기 위해서다. 현실 데이터에 더 안전한 방법이다.

---

### 2️⃣ t 값 = -0.71 의 의미

t 값은

> **평균 차이 ÷ (우연히 생길 수 있는 오차 크기)**

이다.

* 차이: -3.63점
* 오차(표본 변동성): 큼

그래서

$$
t = -0.71
$$

이건 말로 하면:

> 평균 차이가 “우연히 생길 수 있는 흔들림 크기”보다 **작다**

보통 “확실한 차이”라고 하려면 |t|가 2 이상쯤 나오는 경우가 많다.
0.71은 **아주 작은 값**이다.

---

### 3️⃣ p 값 = 0.49 의 의미

p값은 다음 확률이다:

> 두 집단이 원래 **같은 평균**이라고 가정했을 때,
> 지금처럼 3.6점 정도 차이가 **우연히 나올 확률**

결과:

$$
p \approx 0.49 = 49%
$$

해석:

> 이런 차이는 **거의 절반의 경우에 그냥 우연히도 나올 수 있다**

보통 기준 5%(0.05)보다 훨씬 크다 →
**통계적으로 유의하지 않다**.

---

### 4️⃣ 95% 신뢰구간: [-14.27, 7.01]

이건 아주 직관적이다.

> “진짜 평균 차이(추가신청1 − 추가신청0)”가
> 들어 있을 법한 구간

가능한 실제 차이 후보:

* **-14점**일 수도 있음 (추가신청 집단이 훨씬 나쁨)
* **0점**일 수도 있음 (차이 없음)
* **+7점**일 수도 있음 (오히려 더 좋을 수도 있음)

즉,

> **차이 없음(0)** 이 구간 안에 들어 있음 →
> “차이가 있다”는 결론을 낼 수 없다

---

### 5️⃣ 효과크기 Cohen’s d = -0.21

이건 "통계적 유의성"이 아니라 **차이의 크기**를 보는 지표다.

공식 개념:

$$
d = \frac{\text{평균 차이}}{\text{점수의 전반적 흩어짐}}
$$

기준:

| d 값 | 해석    |
| --- | ----- |
| 0.2 | 작은 차이 |
| 0.5 | 중간 차이 |
| 0.8 | 큰 차이  |

결과:

$$
d = -0.21
$$

해석:

> 추가신청 집단이 낮긴 한데,
> **학생들 점수의 자연스러운 흔들림에 비해 매우 작은 차이**

즉, **교육적으로도 큰 차이로 보긴 어려움**.

---

### 🔥 전체 결론 한 줄

> 추가신청 수강생 평균이 조금 낮아 보이지만,
> 그 차이는 **우연으로도 쉽게 생길 수준**이고,
> 실제 차이가 있더라도 **매우 작은 차이**일 가능성이 크다.

즉,

**"추가신청이라서 성적이 낮다"는 증거는 이 데이터로는 부족하다.**

## 4. 중간/기말 성적 변화 분석

</div>

## ChatGPT 설명

중간고사→기말고사 **성적 변화**는 $$\Delta=\text{기말}-\text{중간}$$ 로 정의해 분석한다.

---

### 1) 변화의 전체 규모: “평균은 거의 0, 개인별 변동은 큼”

| 지표           |      값 |
| ------------ | -----: |
| 표본수          |     73 |
| 평균 증감(기말-중간) |   0.90 |
| 중앙값 증감       |   0.00 |
| 표준편차(증감)     |   9.24 |
| IQR(증감)      |  13.50 |
| 최솟값(증감)      | -21.00 |
| 최댓값(증감)      |  30.00 |

핵심 해석

* **평균 증감 0.90점, 중앙값 0점** → 반 전체로 보면 “기말이 중간보다 조금 높아 보이지만 거의 비슷”하다.
* 그런데 **표준편차 9.24점, 범위 -21~+30** → 개인별로는 오르거나 내린 폭이 꽤 크다. 즉, “전체 평균”이 개인별 변화를 가려버리는 상황이다.

---

### 2) 오른 학생 vs 내려간 학생: 반반

| 구분             | 인원 |    비율 |
| -------------- | -: | ----: |
| 상승($\Delta>0$) | 34 | 46.6% |
| 동일($\Delta=0$) |  3 |  4.1% |
| 하락($\Delta<0$) | 36 | 49.3% |

* 상승과 하락이 거의 **1:1**이다.
* 그래서 평균이 0 근처로 수렴한다(오른 사람과 내린 사람이 서로 상쇄).

---

### 3) 변화가 “통계적으로” 의미 있는가: 반 전체 평균 상승은 확실하다고 말하기 어려움

(쌍체 비교: 같은 학생의 두 시험 점수 차이를 직접 본다)

* paired t ≈ **0.836**, p ≈ **0.406**
* 평균 증감 95% 신뢰구간: **[-1.25, 3.06]**
* 효과크기(Cohen’s $d_z$) ≈ **0.098** (매우 작음)

해석

* p=0.406은 “반 전체 평균이 0보다 크다(기말이 더 높다)”를 강하게 지지하지 않는다.
* 신뢰구간에 0이 포함된다 → “진짜 평균 변화가 0일 수도 있다.”
* $d_z$가 0.1 수준 → 반 전체 평균 변화는 **있어도 매우 작다**.

---

### 4) 변화의 분포: 작은 변화가 다수, 큰 변화가 일부

증감 구간별 빈도(명)

| 증감 구간(점)  | 인원 |
| --------- | -: |
| <= -20    |  1 |
| -20 ~ -10 |  4 |
| -10 ~ -5  | 16 |
| -5 ~ 0    | 15 |
| 0 ~ 5     | 17 |
| 5 ~ 10    |  6 |
| 10 ~ 20   | 12 |
| >= 20     |  2 |

관찰

* 가장 많은 구간은 **0~5점 상승(17명)**, 그 다음이 **-10~-5점 하락(16명)**.
* **10점 이상 큰 상승(>=10)** 도 **14명(12+2)** 으로 적지 않다 → “기말에서 크게 회복한 학생”이 눈에 띈다.

---

### 5) 중간 점수와 변화의 관계: 중간이 낮을수록 더 오르는 경향

* corr(중간, 증감) = **-0.551** (상당히 뚜렷한 음의 상관)

해석

* 중간 점수가 **낮았던 학생이 기말에서 더 많이 오르는 경향**이 있다.
* 반대로 중간이 매우 높았던 학생은 기말에서 **유지 또는 소폭 하락**이 섞인다.
* 이는 전형적으로 다음 둘이 함께 나타날 때 자주 보이는 패턴이다.

  * **평균회귀(regression to the mean)**: 극단값은 다음 측정에서 덜 극단적으로 돌아오는 경향
  * **학습 보완**: 중간 이후 학습으로 실제로 성취가 오르는 효과

(이건 방금 생성된 **Midterm vs Score Change** 산점도에서, 왼쪽(중간 낮음)에서 $\Delta>0$ 점들이 많고 오른쪽(중간 높음)에서 $\Delta<0$가 늘어나는 모양으로 확인된다.)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# -----------------------------
# 0) 파생변수
# -----------------------------

df["총점"] = (df["중간고사"] + df["기말고사"]) / 2
df["증감"] = df["기말고사"] - df["중간고사"]

# -----------------------------
# 1) 전체 요약 통계 (원래 형태)
# -----------------------------
cols = ["중간고사", "기말고사", "총점", "출석", "관심도", "증감"]

print("전체 요약 통계\n")
print(df[cols].describe())   # pandas 기본 요약 출력


# -----------------------------
# 2) 상관(핵심만)
# -----------------------------
print("\n상관(핵심 부분)\n")

corr = df[["중간고사", "기말고사", "총점", "증감", "출석", "관심도"]].corr()

corr_focus = corr.loc[
    ["중간고사", "기말고사", "출석", "관심도"],
    ["기말고사", "총점", "증감"]
]

print(corr_focus)   # corr() 기본 출력

# -----------------------------
# 3) 집단 비교: 총점 기술통계 (원래 형태)
# -----------------------------
print("\n성별 총점 요약\n")
print(df.groupby("성별")["총점"].describe())

print("\n학년별 총점 요약\n")
print(df.groupby("학년")["총점"].describe())

print("\n추가신청별 총점 요약\n")
print(df.groupby("추가신청")["총점"].describe())

# -----------------------------
# 4) 극단값(총점 상위 5 / 하위 5)
# -----------------------------
show_cols = ["ID", "학과", "학년", "성별", "중간고사", "기말고사", "출석", "관심도", "총점"]

print("\n총점 상위 5\n")
print(
    df.sort_values("총점", ascending=False)
      .head(5)[show_cols]
)

print("\n총점 하위 5\n")
print(
    df.sort_values("총점", ascending=True)
      .head(5)[show_cols]
)

# -----------------------------
# 5) 그림 3개
# -----------------------------
# (1) 총점 분포 히스토그램
plt.figure()
plt.hist(df["총점"], bins=12)
plt.xlabel("Total score (average of midterm and final)")
plt.ylabel("Number of students")
plt.title("Distribution of Total Score")
plt.tight_layout()
plt.show()

# (2) 중간 vs 기말 산점도 + y=x
plt.figure()
plt.scatter(df["중간고사"], df["기말고사"])
mn = min(df["중간고사"].min(), df["기말고사"].min())
mx = max(df["중간고사"].max(), df["기말고사"].max())
plt.plot([mn, mx], [mn, mx])
plt.xlabel("Midterm score")
plt.ylabel("Final score")
plt.title("Midterm vs Final")
plt.tight_layout()
plt.show()

# (3) 출석 vs 총점 산점도
plt.figure()
plt.scatter(df["출석"], df["총점"])
plt.xlabel("Attendance")
plt.ylabel("Total score")
plt.title("Attendance vs Total Score")
plt.tight_layout()
plt.show()

# -----------------------------
# 6) 추가신청(0 vs 1) Welch t-test, 95% CI, Cohen's d
# -----------------------------
x0 = df.loc[df["추가신청"] == 0, "총점"].astype(float).to_numpy()
x1 = df.loc[df["추가신청"] == 1, "총점"].astype(float).to_numpy()

# Welch t-test
t_stat, p_val = stats.ttest_ind(x1, x0, equal_var=False)  # (1 - 0) 차이 방향

# Welch df (자유도)
n1, n0 = len(x1), len(x0)
m1, m0 = x1.mean(), x0.mean()
s1, s0 = x1.var(ddof=1), x0.var(ddof=1)

se = np.sqrt(s1/n1 + s0/n0)
df_welch = (s1/n1 + s0/n0)**2 / ((s1/n1)**2/(n1-1) + (s0/n0)**2/(n0-1))

diff = m1 - m0  # (추가신청=1) - (추가신청=0)

# 95% CI for mean difference (Welch)
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df_welch)
ci_low = diff - t_crit * se
ci_high = diff + t_crit * se

# Cohen's d (pooled SD 기준; 보고용으로 흔히 이렇게 계산)
sp = np.sqrt(((n1 - 1)*s1 + (n0 - 1)*s0) / (n1 + n0 - 2))
cohen_d = diff / sp

print("\n추가신청 집단 비교(총점): Welch t-test")
print(f"n(추가신청=0)={n0}, mean0={m0:.2f}, sd0={np.sqrt(s0):.2f}")
print(f"n(추가신청=1)={n1}, mean1={m1:.2f}, sd1={np.sqrt(s1):.2f}")
print(f"diff(mean1-mean0)={diff:.2f}")
print(f"t={t_stat:.2f}, df≈{df_welch:.2f}, p={p_val:.3f}")
print(f"95% CI for diff: [{ci_low:.2f}, {ci_high:.2f}]")
print(f"Cohen's d (pooled SD): {cohen_d:.2f}")
```

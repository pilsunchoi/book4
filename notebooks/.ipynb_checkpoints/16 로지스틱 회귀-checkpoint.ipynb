## ChatGPT 설명

다음 데이터는 UCI의 **Bank Marketing Dataset** 중 `bank-additional-full.csv`이다.(데이터 출처: [Bank Marketing Data Set (UCI Machine Learning Repository)](https://archive.ics.uci.edu/dataset/222/bank+marketing)) 

포르투갈의 한 은행이 **정기예금(term deposit) 가입을 유도하기 위해 실시한 전화 마케팅 캠페인 결과**를 기록한 데이터다.

### 1. 데이터 개요

* 관측 수: **41,188**
* 변수 수: **20개 설명변수 + 1개 목표변수**
* 기간: **2008년 5월 ~ 2010년 11월**
* 분석 목적
  → 고객이 **정기예금에 가입할지 여부 예측 (이진 분류 문제)**

목표변수:

### 4. 분석 관점에서 중요한 특징

1. **불균형 데이터**

   * `no`가 대부분
   * → Accuracy보다 **ROC, F1, Precision-Recall**이 중요

2. **범주형 변수 다수**

   * 원-핫 인코딩 필요

    > **원-핫 인코딩(one-hot encoding):** 범주형 변수를 여러 개의 0/1 더미 변수로 바꾸는 방법이다. 각 범주마다 하나의 변수를 만들고, 해당 관측이 속한 범주에는 1, 나머지는 0을 부여한다. 이때 여러 변수 중 하나만 1이 되는(하나만 켜진) 형태이기 때문에 ‘원-핫’이라고 부른다. 예를 들어 직업이 {admin, student, technician}이면 세 개의 더미 변수가 생성된다.
    
3. **정보 누출 가능 변수**

   * `duration` (실제 예측에서는 제외 권장)

4. **마케팅 전략 분석 가능**

   * 연락 횟수 vs 성공률
   * 이전 캠페인 효과
   * 고객 특성별 타겟팅

## ChatGPT 분석

사용한 변수(원변수 기준)

* 수치형: `age`, `campaign`, `previous`, `euribor3m`, `nr.employed`, `cons.conf.idx`, `has_prev_contact`(파생: `pdays != 999`)
* 범주형(더미화): `contact`, `month`, `poutcome`, `education`, `housing`, `loan`

---

### 1) 로지스틱 회귀가 뭐냐 (대학교 저학년/비전공자 버전)

#### (1) 하고 싶은 일

가입 여부 $Y$는 0/1이라서, 선형회귀처럼 $Y$를 그대로 평균으로 예측하면 예측값이 1보다 커지거나 0보다 작아질 수 있다.
그래서 “가입할 확률” $p = P(Y=1\mid X)$ 자체를 예측하되, 항상 0~1 사이가 되게 만드는 모델이 로지스틱 회귀다.

#### (2) 핵심 수식(개념)

로지스틱 회귀는 아래처럼 “확률 $p$”를 바로 직선으로 만들지 않고, **log-odds(로그 오즈)** 를 직선으로 만든다.

* 오즈(odds): $\dfrac{p}{1-p}$
* 로그 오즈(log-odds, logit): $\log\left(\dfrac{p}{1-p}\right)$

로지스틱 회귀 가정:
$$\log\left(\frac{p}{1-p}\right)=\beta_0 + \beta_1 X_1 + \cdots + \beta_k X_k$$

여기서 중요한 해석 포인트는

* 계수 $\beta$는 “확률”을 직접 0.1 올린다 이런 의미가 아니라,
* **오즈(odds)를 몇 배로 바꾸는지**로 해석하는 게 깔끔하다.

#### (3) 오즈비(odds ratio) 해석

어떤 변수 $X_j$가 1 증가할 때,

* 오즈는 $\exp(\beta_j)$배가 된다.
* 그래서 $\exp(\beta_j)$를 **odds_ratio**(오즈비)라고 부른다.

예:

* odds_ratio = 0.95 → 오즈가 0.95배(약 5% 감소)
* odds_ratio = 1.50 → 오즈가 1.5배(50% 증가)

### 3) 결과를 “말로” 읽는 법 (중요한 것부터)

표에서 우선 볼 건 `p_value`가 작은 변수들(통계적으로 “효과가 있다”고 볼 근거가 강한 변수들)이다. 그 다음 **odds_ratio**로 “방향/크기”를 본다.

#### (1) `campaign` (이번 캠페인에서 연락 횟수)

* coef = -0.0516, odds_ratio = 0.95
* 해석: 연락을 1번 더 할수록 “가입 오즈”가 **약 0.95배**(대략 5% 감소)
* 직관: 많이 전화하는 고객이 “원래 거절 성향”이 강해서 더 많이 시도되는 경우가 흔함(선택/운영 전략이 섞여 있을 수 있음).

#### (2) `contact_telephone` (전화기 vs 휴대폰: 기준은 cellular)

* odds_ratio = 0.606
* 해석: `telephone`로 연락한 경우 가입 오즈가 **cellular 대비 약 0.61배**(약 39% 감소)

#### (3) `has_prev_contact` (이전 캠페인 접촉 경험 여부)

* odds_ratio = 3.60
* 해석: 이전에 접촉한 기록이 있는 고객은(= `pdays != 999`) 가입 오즈가 **약 3.6배**
* 직관: 이미 접점이 있는 고객이 더 “따라올” 가능성이 큼.

#### (4) `poutcome_success` (이전 캠페인 성공 경험: 기준은 failure)

* odds_ratio = 3.47
* 해석: 이전 캠페인에서 성공했던 고객은 가입 오즈가 **약 3.5배**

#### (5) `month` 더미들

## 3. 혼동행렬

</div>

## ChatGPT 설명

### 1. 혼동행렬이란?

혼동행렬(confusion matrix)은
**모델의 예측 결과와 실제 결과를 비교한 표**이다.



구성

* **TP (True Positive)**: 실제 Yes → Yes로 맞춤
* **FN (False Negative)**: 실제 Yes → No로 예측 (놓침)
* **FP (False Positive)**: 실제 No → Yes로 예측 (헛예측)
* **TN (True Negative)**: 실제 No → No로 맞춤

즉,

* 위 행: 실제 Yes 중 맞춘 것 vs 놓친 것
* 아래 행: 실제 No 중 맞춘 것 vs 잘못 Yes로 본 것

---

### 2. 주요 성능 지표

혼동행렬의 네 값(TP, FP, FN, TN)으로 여러 성능 지표를 계산한다.

#### (1) 정확도 (Accuracy)

전체 중에서 맞춘 비율
$$
Accuracy = \frac{TP + TN}{TP + FP + FN + TN}
$$

→ 모델이 전반적으로 얼마나 맞는가

---

#### (2) 정밀도 (Precision)

Yes라고 예측한 것 중 실제 Yes 비율
$$
Precision = \frac{TP}{TP + FP}
$$

→ 헛예측이 얼마나 적은가

---

#### (3) 재현율 (Recall, Sensitivity)

실제 Yes 중에서 잡아낸 비율
$$
Recall = \frac{TP}{TP + FN}
$$

→ 중요한 대상을 얼마나 놓치지 않았는가

---

#### (4) F1 점수

Precision과 Recall의 균형 지표
$$
F1 = \frac{2 \cdot Precision \cdot Recall}{Precision + Recall}
$$

---

### 3. 핵심 정리

* Accuracy: 전체 성능
* Precision: 헛예측 관리
* Recall: 놓침 관리
* F1: 두 가지 균형

특히 가입 예측 같은 문제에서는
**Recall(고객을 놓치지 않는 것)**과
**Precision(헛타겟 비용)** 사이의 균형이 중요하다.

## 4. ROC 곡선과 AUC

</div>

## ChatGPT 설명

### ROC 곡선이란?



ROC 곡선은 **모델이 Yes(양성)와 No(음성)를 얼마나 잘 구분하는지**를 보여주는 그래프이다.

* 세로축 (TPR, True Positive Rate)
  → 실제 Yes 중에서 Yes로 맞춘 비율 (Recall)

* 가로축 (FPR, False Positive Rate)
  → 실제 No 중에서 Yes로 잘못 예측한 비율

로지스틱 회귀는 각 관측치에 대해 “Yes일 확률”을 주는데,
**Yes로 판단하는 기준(임계값)** 을 높이거나 낮추면

* Yes를 더 많이 잡을 수도 있고
* 대신 잘못 Yes로 예측하는 경우도 늘어난다.

ROC 곡선은 이 기준을 계속 바꿔가며 나타낸 것이다.

핵심 해석
→ 곡선이 **왼쪽 위에 가까울수록 좋은 모델**이다.

---

### AUC란?

AUC(Area Under the Curve)는
**ROC 곡선 아래 면적**이다.

값의 의미

* 0.5 : 랜덤 추측 수준
* 0.7 ~ 0.8 : 괜찮은 모델
* 0.8 ~ 0.9 : 좋은 모델
* 1.0 : 완벽한 구분

직관적 의미
→ 임의로 Yes 1명과 No 1명을 뽑았을 때,
모델이 Yes에게 더 높은 확률을 줄 확률

---

### 한 줄 정리

ROC는 **구분 능력을 그림으로 보여주는 것**,
AUC는 **그 구분 능력을 하나의 숫자로 요약한 것**이다.

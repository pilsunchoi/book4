### 2. 시각화 분석 (Python Code)

데이터를 바탕으로 연령별, 지역별 변화 추이를 시각화하겠습니다.

두 시점(2026년 1월 2주 vs 2월 2주) 간의 대통령 지지율 변화에 대한 분석 결과입니다.

#### 1. 종합 요약

* **전체 지지율 상승:** 대통령 직무 수행 긍정 평가는 1월 2주 **60%** 에서 2월 2주 **63%** 로 **3%포인트 상승**했습니다.
* **30대와 영남권의 결집:** 이번 지지율 상승은 주로 **30대** 연령층과 **대구/경북, 부산/울산/경남** 지역의 지지율 급등이 견인한 것으로 분석됩니다.
* **20대와 50대의 이탈:** 반면, 18~29세와 50대에서는 지지율이 하락하는 추세를 보였습니다.

#### 2. 연령별 변화 분석






| 연령대 | 1월 2주 (%) | 2월 2주 (%) | 변화 (%p) | 특징 |
| --- | --- | --- | --- | --- |
| **18~29세** | 46 | 39 | ▼ **-7** | **가장 큰 하락폭.** 젊은 층의 지지 이탈이 두드러짐. |
| **30대** | 50 | 66 | ▲ **+16** | **폭발적 상승.** 전 연령대 중 가장 큰 상승폭 기록. |
| **40대** | 75 | 75 | - | 변화 없음. 견고한 지지층 유지. |
| **50대** | 77 | 70 | ▼ **-7** | 20대와 함께 하락세 주도. |
| **60대** | 56 | 65 | ▲ **+9** | 지지율 회복세 뚜렷. |
| **70대 이상** | 49 | 57 | ▲ **+8** | 과반 지지율 회복. |

* **분석:** 30대에서의 지지율이 16%p나 급등한 것이 전체 지지율 상승의 핵심 요인입니다. 반면, 20대(18~29세)와 50대에서는 각각 7%p씩 하락하며 세대별로 지지율 변화의 방향이 엇갈리는 현상이 나타났습니다.

#### 3. 지역별 변화 분석





| 지역 | 1월 2주 (%) | 2월 2주 (%) | 변화 (%p) | 특징 |
| --- | --- | --- | --- | --- |
| **서울** | 59 | 58 | ▼ -1 | 소폭 하락, 보합세. |
| **인천/경기** | 65 | 62 | ▼ -3 | 소폭 하락. |
| **대전/세종/충청** | 58 | 69 | ▲ **+11** | **충청권 급등.** |
| **광주/전라** | 82 | 81 | ▼ -1 | 압도적 지지율 유지 중이나 소폭 조정. |
| **대구/경북** | 39 | 49 | ▲ **+10** | **큰 폭 상승.** 긍정 평가가 부정 평가를 상회하기 시작. |
| **부산/울산/경남** | 49 | 63 | ▲ **+14** | **최대 상승 지역.** 과반 지지율 확보. |

* **분석:** 수도권(서울, 인천/경기)과 호남권은 약보합세를 보였으나, **영남권(대구/경북, 부산/울산/경남)과 충청권**에서 지지율이 10%p 이상 급등하며 전체 지지율 상승을 이끌었습니다. 특히 부산/울산/경남 지역의 14%p 상승은 매우 주목할 만한 변화입니다.

이 데이터는 제공해주신 파일 내의 수치를 기반으로 재구성한 것이며, 표본오차(±3.1%p)를 고려할 때 30대와 영남/충청권의 상승세는 통계적으로도 유의미한 변화일 가능성이 높습니다.

## 2. 젊은층(18~29세)의 지지 이탈은 통계적으로 유의한가?

</div>

## Gemini 분석

두 시점(2026년 1월 2주 vs 2월 2주) 간 **18~29세(젊은 층)** 의 대통령 지지율 변화에 대한 통계적 검정 결과입니다.

### 분석 결과 요약

* **1월 2주 지지율:** 46% (표본 수: 142명) 


* **2월 2주 지지율:** 39% (표본 수: 133명) 


* **지지율 차이:** -7%포인트 (46% → 39%)

---

### 통계적 유의성 검정 (두 비율 차이 검정, Two-proportion z-test)

* **Z-score:** 1.2466
* **P-value (유의확률):** 0.2125

---

### 결론

통계적으로 **유의한 차이가 아닙니다.** (유의수준 5% 기준, P-value > 0.05)

즉, 18~29세 그룹에서 관측된 **7%포인트의 지지율 하락**은 표본 오차 범위 내의 변동일 가능성이 높으며, 이를 **통계적으로 확실한 지지 이탈이라고 단정하기 어렵습니다.**

이는 해당 연령대의 표본 수(약 130~140명)가 전체 표본(약 1,000명)에 비해 상대적으로 적기 때문에, 오차 범위가 넓게 형성되어 나타나는 결과입니다. 따라서 이 변화를 확정적인 추세로 해석하기보다는 **'하락세가 관측되나 추가적인 확인이 필요하다'** 정도로 이해하는 것이 타당합니다.

## 3. 다음 표현에 대해 알기쉽게 설명해줘

<h4 style="text-align: center;">
  "통계적으로 유의한 차이가 아닙니다. (유의수준 5% 기준, P-value > 0.05)"
</h4>

</div>

## Gemini 분석

**"통계적으로 유의한 차이가 아니다 (P-value > 0.05)"** 라는 말은, 쉽게 말해 **"이번 조사에서 숫자가 떨어진 것처럼 보이지만, 이게 진짜 실력이 아니라 '우연'일 수도 있다"** 는 뜻입니다.

비유를 들어 더 쉽게 설명해 드릴게요.

### 1. 핵심 의미: "우연일지도 몰라요"

여론조사는 전 국민을 다 조사하는 게 아니라, 그중 일부(샘플)만 뽑아서 조사합니다. 그러다 보니 **누가 전화를 받느냐에 따라** 숫자가 조금씩 오르락내리락할 수 있습니다.

* **통계적으로 유의하다:** "이건 우연히 나올 수 없는 숫자다. 진짜로 지지율이 변했다." (확신 O)
* **통계적으로 유의하지 않다:** "누가 전화를 받느냐에 따라 우연히 이렇게 나왔을 수도 있다. 진짜로 변했는지는 확신할 수 없다." (확신 X)

---

### 2. 숫자로 보는 기준 (P-value)

통계학에서는 '우연일 확률'을 계산하는데, 이걸 **P-value(유의확률)** 라고 부릅니다.

* **기준점 (0.05 = 5%):** 우리는 보통 **"우연일 확률이 5% 미만이어야 진짜라고 믿자"** 라고 약속합니다. 100번 조사했을 때 95번 이상은 똑같은 결과가 나와야 믿겠다는 거죠.
* **이번 결과 (0.21 = 21%):** 아까 계산해보니 P-value가 **0.21**이 나왔습니다. 이건 **"이 결과가 우연히 나왔을 확률이 21%나 된다"** 는 뜻입니다.

**결론:** 우연일 가능성이 5%보다 훨씬 높으니(21% > 5%), 통계학적으로는 **"이 하락폭(-7%p)은 진짜 하락세라고 단정 짓기엔 증거가 부족하다"** 라고 판단하는 것입니다.

---

### 3. 왜 이런 결과가 나왔나요?

가장 큰 이유는 **조사 대상자 수(표본 수)** 가 적기 때문입니다.

* 이번 20대 응답자는 약 130~140명이었습니다.
* 사람 수가 적으면 한두 명만 마음을 바꿔도 퍼센트(%)가 확 뜁니다.
* 그래서 7%포인트나 떨어졌음에도 불구하고, 통계적으로는 "이 정도 인원수에서는 오차 범위 안에서 흔히 일어날 수 있는 일"이라고 보는 것입니다.

**요약하자면:**
젊은 층의 지지율이 떨어진 것**처럼 보이지만**, 다음 주에 다시 조사하면 도로 올라갈 수도 있는, **아직은 확신할 수 없는 변화**라는 뜻입니다.

## 참고: 파이썬 코드

</div>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 데이터 정의
data_age = {
    'Age Group': ['18-29', '30s', '40s', '50s', '60s', '70+'],
    'Jan 2nd Week': [46, 50, 75, 77, 56, 49],
    'Feb 2nd Week': [39, 66, 75, 70, 65, 57]
}

data_region = {
    'Region': ['Seoul', 'Incheon/Gyeonggi', 'Daejeon/Chungcheong', 
               'Gwangju/Jeonla', 'Daegu/Gyeongbuk', 'Busan/Ulsan/Gyeongnam'],
    'Jan 2nd Week': [59, 65, 58, 82, 39, 49],
    'Feb 2nd Week': [58, 62, 69, 81, 49, 63]
}

df_age = pd.DataFrame(data_age)
df_region = pd.DataFrame(data_region)

# 변동폭 계산
df_age['Change'] = df_age['Feb 2nd Week'] - df_age['Jan 2nd Week']
df_region['Change'] = df_region['Feb 2nd Week'] - df_region['Jan 2nd Week']

# 1. 연령별 지지율 비교 (Bar Chart)
plt.figure(figsize=(6, 4))
x = np.arange(len(df_age['Age Group']))
width = 0.35

rects1 = plt.bar(x - width/2, df_age['Jan 2nd Week'], width, label='Jan 2nd Week', color='#A9A9A9')
rects2 = plt.bar(x + width/2, df_age['Feb 2nd Week'], width, label='Feb 2nd Week', color='#1f77b4')

plt.ylabel('Approval Rating (%)')
plt.title('Presidential Approval Rating by Age Group')
plt.xticks(x, df_age['Age Group'])
plt.legend()
plt.ylim(0, 90)
plt.bar_label(rects1, padding=3)
plt.bar_label(rects2, padding=3)
plt.tight_layout()
plt.show()

# 2. 연령별 지지율 변화 (Diverging Bar Chart)
plt.figure(figsize=(6, 4))
colors_age = ['red' if x < 0 else 'green' for x in df_age['Change']]
container = plt.barh(df_age['Age Group'], df_age['Change'], color=colors_age)
plt.title('Change in Approval Rating by Age (Jan -> Feb)')
plt.xlabel('Percentage Point Change (%p)')
plt.axvline(0, color='black', linewidth=0.8)
plt.bar_label(container, padding=3)
plt.tight_layout()
plt.show()

# 3. 지역별 지지율 비교 (Bar Chart)
plt.figure(figsize=(8, 4))
x_reg = np.arange(len(df_region['Region']))

rects3 = plt.bar(x_reg - width/2, df_region['Jan 2nd Week'], width, label='Jan 2nd Week', color='#A9A9A9')
rects4 = plt.bar(x_reg + width/2, df_region['Feb 2nd Week'], width, label='Feb 2nd Week', color='#ff7f0e')

plt.ylabel('Approval Rating (%)')
plt.title('Presidential Approval Rating by Region')
plt.xticks(x_reg, df_region['Region'], rotation=15, ha='right')
plt.legend()
plt.ylim(0, 100)
plt.bar_label(rects3, padding=3)
plt.bar_label(rects4, padding=3)
plt.tight_layout()
plt.show()

# 4. 지역별 지지율 변화 (Diverging Bar Chart)
plt.figure(figsize=(7, 4))
colors_reg = ['red' if x < 0 else 'green' for x in df_region['Change']]
container = plt.barh(df_region['Region'], df_region['Change'], color=colors_reg)
plt.title('Change in Approval Rating by Region (Jan -> Feb)')
plt.xlabel('Percentage Point Change (%p)')
plt.axvline(0, color='black', linewidth=0.8)
plt.bar_label(container, padding=3)
plt.tight_layout()
plt.show()
```

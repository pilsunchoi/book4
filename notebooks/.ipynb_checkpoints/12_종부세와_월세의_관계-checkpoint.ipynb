# **12장 종부세와 월세의 관계**

## <span style="color: blue;">**종부세가 월세를 높였다?**</span>

![Rent](https://raw.githubusercontent.com/pilsunchoi/images4/main/12-1.png)

<span style="color: blue;">종합부동산세(종부세) 등 부동산 보유세를 늘리면 월세가 급등한다는 실증 분석 결과가 나왔다. 집주인들이 보유세를 임차인에게 전가하는 효과가 나타난다는 것이다. 분석 결과에 따르면 종부세가 도입된 노무현 정부(2003-2008년) 시기에는 월세가 20% 넘게 올랐다. 종부세가 확대·강화된 문재인 정부(2017-2022년) 시기에도 30% 넘는 월세의 동반 상승이 나타났다.</span>

<span style="color: blue;">종부세는 노무현 정부 시기인 2005년 1월 5일 제정됐고, 이명박 정부가 들어선 2008년 9월 1일 '9·1 세제 개편안'에 따라 세 부담이 완화됐다. 그러나 문재인 정부가 집권한 이후 2018년 9월 13일 '주택시장안정화대책', 2019년 12월 16일 '주택시장 안정화 방안' 등을 거치며 다시 세율이 높아졌다.</span>
 
<span style="color: blue;">28일 학계에 따르면 이창무 한양대 교수 연구실은 27일 '이재명 정부 초기 부동산시장 현황 및 정책 진단' 연구를 국회에 공개했다. 이 연구는 종부세의 월세 전가 효과를 알아보기 위해 2005년부터 2022년까지 서울·경기·인천의 종부세 징수액과 2006년부터 2023년까지 월세 지수 및 월세 시계열 자료를 분석했다.</span>
 
<span style="color: blue;">종부세 징수액과 월세의 관계만을 알아보기 위해 주택 공급 및 수요, 거시경제 등은 통제 변수로 영향을 배제했다. 또 월세 수준은 부동산R114의 월세 지수와 국토부에 등록된 실거래가를 통해 자체 산정한 실거래가 지수 2가지로 별도 측정했다. 부동산R114 월세지수와 실거래가 지수 모두 2011년 1월 월세를 기준치인 100으로 보고 상대적 수준을 측정한 지표다. 다만 종부세 징수액 통계를 2022년 말까지만 사용해 이 이후의 월세 상승률에 대해서는 분석하지 못한 한계가 있다.</span>

<span style="color: blue;">이 교수에 따르면 도입 원년인 2005년 서울시의 종부세 징수액은 289억원이었지만 2006년에는 3824억원까지 늘었고 2007년에는 9037억원으로 1조원에 육박했다. 2008년에는 6107억원이 징수됐다. 2008년 종부세 징수액 감소는 글로벌 금융 위기와 이명박 정부의 세액 감소 정책(9·1 세제 개편안)의 단계적 적용 등의 영향이다. 2005년부터 2008년까지 총징수액은 1조9259억원이다.</span>
 
<span style="color: blue;">서울의 부동산R114 월세 지수는 2005년 1월 79.80이었다. 그러나 2006년 12월에는 80.70이 됐고 2007년 12월에는 90.05로 올랐다. 또 이듬해인 2008년 말에는 96.18까지 상승했다. 이 기간 지수 상승률은 20.5%(16.38포인트)이다.</span>
 
<span style="color: blue;">같은 기간 경기도 종부세 징수액도 2005년 60억원에서 2006년 1073억원으로 늘었고 2007년에는 2849억원이, 2008년에는 1650억원이 징수됐다. 총징수액은 5632억원이다. 이 기간 경기도의 부동산R114 월세 지수도 2005년 1월 82.91에서 2008년 말 97.73으로 17.8%(14.82포인트) 상승했다.</span>
 
<span style="color: blue;">노무현 정부 시절 종부세 도입과 강화 이후 서울, 경기 월세는 20%가량 상승한 셈이다. 이창무 교수는 "월세 거래 가격을 분석, 판단한 결과 종부세가 월세 급등으로 이어지는 움직임이 드러난다"고 설명했다.</span>
 
<span style="color: blue;">문재인 정부에서도 비슷한 양상이 나타났다. 문 정부는 2017년 5월 출범해 '9·13 부동산 대책' '12·16 부동산 대책' 등을 내놓으며 종부세 세율 증가 등 규제 강화책을 사용했다. 이런 기조는 윤석열 정부가 들어서 2022년 세제 개편(안)을 확정(2022년 7월 21일)하며 다주택자 종부세율을 인하하고 기본 공제 금액을 상향할 때까지 이어졌다.</span>
 
<span style="color: blue;">문재인 정부 집권 이듬해인 2018년 서울 종부세 부과액은 2754억원이었지만 2019년에는 6193억원으로 증가했고 2020년(9965억원), 2021년(2조3741억원)에도 세수는 계속 늘었다. 이 기간 총 징수액은 4조2655억원이다.</span>
 
<span style="color: blue;">부동산R114 월세 지수는 2018년 1월 103.74로 기준 연도인 2011년 1월보다 소폭 높은 수준이었다. 그러나 종부세 강화 이후인 2021년 12월에는 114.36으로 상승했고 문재인 정부 마지막 해인 2022년에는 123.46까지 올랐다. 이 기간 누적 월세 지수 상승률은 19.0%(19.72포인트)다.</span>

<span style="color: blue;">이 교수 연구실이 국토부 실거래가를 토대로 자체 산정한 실거래가 지수를 봐도 월세 급등은 확인됐다. 2018년 1월 월세 실거래가 지수는 96.39로 기준 연도인 2011년 1월보다 낮은 수준이었다. 그러나 2022년 12월에는 이 지수가 127.32까지 올랐다. 누적 상승률은 32.08%(30.93포인트)다. 2018년 월세 100만원을 내고 살았던 사람이 2022년 말에는 월세 132만원을 내야 했다는 의미다.</span>
 
<span style="color: blue;">이 교수는 "노무현 정부 시기 종부세 도입과 강화는 20%에 가까운 월세 급등을 초래했고 문재인 정부 시기 종부세의 재강화는 실거래가 지수로 30~40%의 월세 급등세를 초래했다"고 밝혔다. 이 교수는 "종부세 징수액을 2022년 말까지 통계를 활용해 분석해 윤석열 정부의 감세안 발표 후 서울 등 일부 지역의 월세 상승 추세를 분석하지는 못했다"고 설명했다. 그는 "다만 최근 월세 상승은 종부세 요인뿐 아니라 정부의 대출 규제 등 다른 규제의 영향이 상승 효과를 만들어 낸 것으로 예상한다"고 했다. 이 연구는 27일 오후 국회 의원회관에서 열린 '부동산 정책 이대로 괜찮은가?' 정책 세미나의 기조 발제로 공개됐다.</span>

<span style="color: blue;">부동산 업계 관계자는 "집주인들이 정부의 보유세 강화 시기에는 세 부담을 세입자에게 전가하는 경향이 강하고 이런 이유로 보유세 강화가 월세 상승으로 이어지는 것"이라고 했다.</span>

## **Claude 평가**

이 연구는 흥미로운 주제를 다루고 있지만, 방법론과 해석 면에서 몇 가지 중요한 한계가 있습니다.

**연구의 강점**

장기 시계열(2005~2022)을 활용하고, 두 가지 월세 지수(부동산R114, 실거래가 기반)를 병행 분석한 점은 긍정적입니다. 주택 공급·수요, 거시경제 변수를 통제했다고 밝힌 점도 기본적인 방법론적 노력이 보입니다.

**핵심적인 문제점들**

첫째, **인과관계 추론의 한계**가 가장 큽니다. 종부세 징수액과 월세가 같은 시기에 올랐다는 것만으로 "종부세가 월세 상승을 초래했다"고 결론 내리기는 어렵습니다. 두 변수 모두 부동산 가격 상승이라는 공통 원인에 의해 동시에 움직일 수 있습니다. 집값이 오르면 종부세 과세 대상과 세액이 늘어나고, 동시에 월세도 오르는 것은 자연스러운 현상이기 때문입니다. 이른바 **허위 상관(spurious correlation)** 가능성이 큽니다.

둘째, **통제 변수의 적절성**이 의문입니다. "주택 공급 및 수요, 거시경제를 통제했다"고 하지만, 구체적으로 어떤 변수를 어떻게 통제했는지 명확하지 않습니다. 저금리 기조, 전세의 월세 전환 추세, 임대차 3법(2020년) 시행 등 월세에 직접적으로 영향을 미치는 핵심 요인들이 제대로 분리되었는지 확인하기 어렵습니다.

셋째, **전세→월세 구조 전환** 효과를 간과하고 있습니다. 특히 문재인 정부 시기 월세 상승은 임대차 3법 시행 이후 전세 공급 감소와 월세 전환 가속화의 영향이 매우 컸습니다. 이 효과를 종부세 효과와 분리하지 않으면 종부세의 영향을 과대추정할 위험이 있습니다.

넷째, **분석 기간 선택의 문제**입니다. 노무현 정부 시기(2005-2008)와 문재인 정부 시기(2018-2022)만 집중 조명하고, 이명박·박근혜 정부 시기에 종부세가 완화되었을 때 월세가 어떻게 변동했는지에 대한 대칭적 분석이 부족합니다. 그래프를 보면 종부세 완화 시기에도 월세가 꾸준히 상승한 것으로 보이는데, 이는 오히려 종부세 외 요인이 월세 상승의 주된 동인임을 시사합니다.

다섯째, 연구자 본인도 인정하듯 **2022년 이후 윤석열 정부의 감세 효과를 검증하지 못했습니다.** 만약 종부세 인하 이후에도 월세가 계속 올랐다면, 이 연구의 핵심 주장이 약화됩니다.

**종합 평가**

이 연구는 "종부세 → 월세 전가"라는 하나의 가설을 제시하는 데는 의미가 있지만, 상관관계를 인과관계로 해석하는 경향이 강하고, 경쟁 가설들(전세-월세 전환, 저금리, 임대차 3법 등)에 대한 충분한 통제와 비교가 부족합니다. 정책 세미나에서 발표된 연구라는 점에서, 특정 정책 방향을 지지하기 위한 목적성이 있을 수 있다는 점도 감안하여 읽을 필요가 있습니다.

## **ChatGPT 평가**

이 분석은 “보유세 증가 → 월세 급등”이라는 **인과관계(causality)** 를 주장하고 있지만, 현재 제시된 방식은 학술적으로 보면 **강한 상관관계 서술에 가까우며, 인과추론 설계로는 상당히 취약**하다. 핵심 쟁점을 계량경제학 관점에서 정리해 보겠다.

---

### 1. 이 연구가 실제로 한 것: 구조

연구는 대략 다음 구조다.

* 설명변수: 서울·경기·인천의 종부세 징수액
* 종속변수: 월세 지수

  * 부동산R114 월세지수
  * 국토교통부 실거래가 기반 자체 지수
* 기간: 2005~2022년
* 통제: 공급·수요·거시 변수 일부 통제했다고 주장

그리고 다음을 근거로 결론을 낸다.

> 종부세가 급증한 시기(노무현·문재인 정부)와 월세 급등 시기가 겹친다 → 세금 전가 발생

겉보기에는 그럴듯하지만, **계량적으로는 가장 위험한 형태의 분석 구조**다.

---

### 2. 가장 큰 문제: “세금이 늘어서 월세가 오른 것”이 아님

#### ❗ 핵심 식별 문제: **내생성(endogeneity)**

종부세 징수액은 정책 변수처럼 보이지만, 실제로는

$$\text{종부세 징수액} = f(\text{세율}, \text{공시가격}, \text{주택가격 수준}, \text{주택 수})$$

즉,

> **집값이 오르면 세금도 자동으로 늘어난다.**

월세도 집값과 같은 부동산 시장 여건의 영향을 받는다.

따라서 관측된 관계는 실제로:

> 부동산 가격 상승 국면 → 종부세 징수액 ↑
> 부동산 가격 상승 국면 → 월세 ↑

이 **공통 원인(common cause)** 때문일 가능성이 매우 크다.

이 경우 추정식은 사실상

$$\text{월세}_t = \beta \cdot \text{종부세징수액}_t + u_t$$

이지만 진짜 구조는

$$
\text{월세}_t = \gamma \cdot \text{부동산경기}_t + \epsilon_t
$$
$$
\text{종부세징수액}_t = \delta \cdot \text{부동산경기}_t + \nu_t
$$

여기서 $\text{부동산경기}_t$가 빠져 있다면, $\beta$는 **심각한 편의(bias)** 를 가진다.

👉 이건 전형적인 **omitted variable bias** 상황이다.

---

### 3. “통제변수 넣었다”로 해결되지 않는 이유

연구는 “공급·수요·거시경제 변수를 통제했다”고 하지만, 다음이 중요하다.

#### 부동산 시장의 핵심 상태변수는 관측 불완전

* 기대 가격 상승률
* 투자 수요 심리
* 전세→월세 전환 압력
* 정책 리스크 기대

이런 것들은 제대로 계량화하기 어렵다.
특히 2017~2022년은 대출규제, 전세대출, 임대차3법 등 구조 변화가 겹친 시기다.

즉, **세금과 동시에 움직인 다른 충격이 너무 많다.**

---

### 4. “징수액”을 쓰는 것 자체가 매우 취약

보통 세금 전가(pass-through) 연구는 다음을 본다.

* 세율 변화
* 과세 기준 변화
* 특정 집단에만 적용되는 정책 변화

그런데 여기서는 **징수액(aggregate tax revenue)** 을 사용했다.

이건 문제가 크다.

> 징수액 = 세율 효과 + 과세표준 증가 효과 + 납세자 수 증가 효과

즉, 정책 때문인지 **집값 상승 때문인지 분리 불가능**하다.

학술적으로는 세율이나 과세 기준 변화 같은 **외생적 정책 쇼크(exogenous policy shock)** 를 써야 한다.

---

### 5. 단순 시계열 상관은 거의 의미가 없다

이 연구는 사실상:

> A 정부 시기 세금 ↑ + 월세 ↑
> B 정부 시기 세금 ↑ + 월세 ↑

라는 **레짐 비교(regime comparison)** 에 가깝다.

하지만 같은 시기에 일어난 것들:

#### 노무현 정부(노무현)

* 집값 급등기
* 수도권 주택 수요 폭증
* 전세시장 구조 변화

#### 문재인 정부(문재인)

* 저금리 + 유동성 폭증
* 임대차3법
* 전세가 폭등 → 월세 전환
* 대출 규제

즉, “종부세가 오른 시기” = “부동산 시장이 격변하던 시기”다.

이걸 분리하지 않으면 **정책 효과가 아니라 시대 효과(period effect)** 를 잡는 것이다.

---

### 6. 진짜 세금 전가 연구라면 이렇게 해야 한다

학술적으로 설득력 있으려면 최소한:

#### ✔ (1) DID(이중차분) 구조 필요

예:

* 종부세 영향 큰 지역 vs 거의 없는 지역
* 과세 대상 주택 vs 비대상 주택

$$
\text{월세}_{it} = \alpha + \beta(\text{과세대상}_i \times \text{정책후}_t) + \text{고정효과} + \epsilon_{it}
$$

이렇게 해야 “세금 맞은 집”과 “안 맞은 집”을 비교할 수 있다.

#### ✔ (2) 세율 변화 중심 분석

징수액이 아니라:

* 세율 인상 폭
* 공제 기준 변화
* 과세표준 구간 이동

같은 **정책 설계 변수**를 써야 한다.

#### ✔ (3) 집값·전세가 동학을 명시적으로 모델링

월세는 전세 시장과 강하게 연결되어 있다.
문재인 정부 시기 월세 급등의 핵심은:

> 전세 공급 감소 → 월세 전환

이 효과를 제거하지 않으면 세금 효과 추정이 왜곡된다.

---

### 7. 결론: 이 연구의 신뢰도

| 평가 항목      | 판단     |
| ---------- | ------ |
| 상관관계 묘사    | 있음     |
| 인과 추정 설계   | 매우 취약  |
| 내생성 처리     | 사실상 없음 |
| 정책 효과 분리   | 실패     |
| 학술 논문 수준   | 미달     |
| 정책 주장용 보고서 | 가능     |

이건 **정책 토론용 주장 자료**로는 쓸 수 있지만,
“종부세 때문에 월세가 20~40% 올랐다”는 **정량적 인과 주장으로는 설득력이 부족**하다.

---

### 한 줄 요약

> 이 분석은 “종부세가 오른 시기와 월세가 오른 시기가 겹쳤다”는 것을 보여줄 뿐,
> “종부세 때문에 월세가 올랐다”는 것을 입증하지는 못한다.

즉, **상관관계는 있으나 인과관계 증거는 약하다**가 정확한 평가다.

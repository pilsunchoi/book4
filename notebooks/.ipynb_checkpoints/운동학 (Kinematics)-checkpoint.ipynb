아래 결과는 업로드된 동영상에서 `OpenCV` 배경차분으로 “사람(전경)”의 가장 큰 윤곽을 추적하고, 그 윤곽의 무게중심(centroid)·바운딩박스·윤곽 주축(PCA) 각도를 이용해 운동학 변수를 “프록시(proxy)”로 계산한 것이다. 절대 단위는 캘리브레이션이 없어서 `px`, `px/s` 같은 픽셀 단위다.

**분석 메모**

* 움직임이 거의 없는 구간이 길어서, (A) 전체 구간 통계와 (B) “활동 프레임(speed > 5 px/s)” 통계를 따로 제시한다.

**해석(전체 구간)**

* `Speed/Acceleration/Jerk`의 median이 0인 것은 “대부분 프레임에서 전경 추적 결과가 거의 정지(또는 전경이 작게 잡힘)”임을 의미한다. 즉, 전체 평균은 “정지 구간 + 움직임 구간”이 섞인 값이다.
* `Displacement from start`의 최대가 약 338.55 px이고, `Path length` 최대가 약 427.11 px이라서, 전체적으로는 “초기 위치 대비 꽤 이동했지만(최대 변위 338 px), 이동 경로는 더 길었다(누적 427 px)” → 방향 전환이나 왕복 성분이 있었다는 신호다.
* `BBox height/width`가 움직임 구간에서 커지는 패턴이 나타났는데(아래 그래프), 이는 실제로 몸이 펴지거나(자세 변화), 카메라에 가까워지거나, 전경 분리가 더 크게 잡히는 상황 중 하나(또는 혼합)일 수 있다.
* `Body axis angle`은 윤곽 주축의 각도라서 “상체 기울기”를 직접 의미하진 않지만, 움직임 구간에서 각도가 변하면 “전신 방향/자세가 변했다”는 정성적 신호로 해석할 수 있다.

**해석(활동 프레임 중심)**

* “움직일 때의 대표 속도”는 median 기준 약 301 px/s이고, 상위 5%는 약 454 px/s 수준이다.
* 가속·저크가 상당히 큰 편으로 나오는 구간이 있는데, 이 값들은 (1) 실제 급가속/급정지, (2) 전경 분리 마스크의 흔들림, (3) 프레임 단위 차분의 노이즈가 섞여 커질 수 있다. 그래서 저크는 “매끈함(smoothness) 비교용”으로는 유용하지만, 절대값 자체를 물리량처럼 해석하는 건 위험하다.
* 움직임 구간에서 `BBox height`가 높게 유지(중앙값 154 px)되는 것은, 최소한 “전경이 크게 잡히는 자세/거리” 상태가 지속됐다는 의미다.

---

### 3) 그래프에서 읽을 수 있는 핵심

* Speed time series: 실제 움직임은 대략 2.6s~4.1s 부근에 집중되어 있고 그 외는 거의 정지로 보인다.


* BBox height time series: 움직임 구간에서 전경의 세로 크기가 급격히 증가했다가 종료 후 감소한다.


* Body axis angle time series: 움직임 구간에서 각도 변동이 커지며, 종료 시점 이후에는 거의 상수처럼 유지(정지/추적 고정)된다.


---

원하면 다음도 같은 방식으로 바로 계산해서 붙일 수 있다.

* “구간 분할(반복 동작 rep 자동 탐지)”을 speed 피크 기반으로 하고, rep별 속도/가속/ROM(프록시) 통계를 따로 뽑기
* 픽셀을 실제 단위(m/s)로 바꾸기(바닥 기준 길이 1개만 알려주면 스케일 캘리브레이션 가능)

## **참고: 파이썬 코드**

```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

video_path = "../Data/22-1.mp4"
assert os.path.exists(video_path), video_path

# -----------------------------
# 1) 비디오 메타데이터
# -----------------------------
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
n_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
cap.release()

print("fps:", fps, "frames:", n_frames, "size:", (w, h))

# -----------------------------
# 2) 전경(사람) 추적: 배경차분 + 최대 윤곽
# -----------------------------
cap = cv2.VideoCapture(video_path)
bg = cv2.createBackgroundSubtractorMOG2(history=200, varThreshold=25, detectShadows=True)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

records = []
frame_idx = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    fg = bg.apply(gray)

    # 그림자 제거: MOG2는 shadow를 127로 주는 경우가 많음
    _, fg_bin = cv2.threshold(fg, 200, 255, cv2.THRESH_BINARY)
    fg_bin = cv2.morphologyEx(fg_bin, cv2.MORPH_OPEN, kernel, iterations=1)
    fg_bin = cv2.morphologyEx(fg_bin, cv2.MORPH_CLOSE, kernel, iterations=2)

    contours, _ = cv2.findContours(fg_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    row = {"frame": frame_idx, "t": frame_idx / fps, "fg_pixels": int(fg_bin.sum() // 255)}
    if contours:
        cnt = max(contours, key=cv2.contourArea)
        area = cv2.contourArea(cnt)

        if area > 150:  # 노이즈 제거
            x, y, bw, bh = cv2.boundingRect(cnt)

            M = cv2.moments(cnt)
            if M["m00"] != 0:
                cx = M["m10"] / M["m00"]
                cy = M["m01"] / M["m00"]
            else:
                cx, cy = np.nan, np.nan

            # 윤곽 주축 방향(PCA) 각도
            pts = cnt.reshape(-1, 2).astype(np.float32)
            angle = np.nan
            if pts.shape[0] >= 5:
                mean = pts.mean(axis=0)
                X = pts - mean
                cov = (X.T @ X) / max(len(X) - 1, 1)
                eigvals, eigvecs = np.linalg.eigh(cov)
                v = eigvecs[:, np.argmax(eigvals)]
                angle = float(np.degrees(np.arctan2(v[1], v[0])))

            row.update({
                "cx": cx, "cy": cy,
                "area": area,
                "bbox_w": bw, "bbox_h": bh,
                "bbox_x": x, "bbox_y": y,
                "angle_deg": angle
            })
        else:
            row.update({"cx": np.nan, "cy": np.nan, "area": np.nan,
                        "bbox_w": np.nan, "bbox_h": np.nan,
                        "bbox_x": np.nan, "bbox_y": np.nan, "angle_deg": np.nan})
    else:
        row.update({"cx": np.nan, "cy": np.nan, "area": np.nan,
                    "bbox_w": np.nan, "bbox_h": np.nan,
                    "bbox_x": np.nan, "bbox_y": np.nan, "angle_deg": np.nan})

    records.append(row)
    frame_idx += 1

cap.release()

df = pd.DataFrame(records)

# 결측 보간(미분 계산 안정화 목적)
df_i = df.copy()
for col in ["cx", "cy", "area", "bbox_w", "bbox_h", "angle_deg"]:
    df_i[col] = df_i[col].interpolate(limit_direction="both")

# -----------------------------
# 3) 운동학 변수(프록시) 계산
# -----------------------------
dt = 1.0 / fps
x = df_i["cx"].to_numpy()
y = df_i["cy"].to_numpy()

vx = np.gradient(x, dt)
vy = np.gradient(y, dt)
speed = np.sqrt(vx * vx + vy * vy)                # px/s

ax = np.gradient(vx, dt)
ay = np.gradient(vy, dt)
acc = np.sqrt(ax * ax + ay * ay)                  # px/s^2

jx = np.gradient(ax, dt)
jy = np.gradient(ay, dt)
jerk = np.sqrt(jx * jx + jy * jy)                 # px/s^3

disp = np.sqrt((x - x[0]) ** 2 + (y - y[0]) ** 2) # px
step = np.sqrt(np.diff(x) ** 2 + np.diff(y) ** 2)
path_len = np.concatenate([[0.0], np.cumsum(step)])  # px

bbox_h = df_i["bbox_h"].to_numpy()
bbox_w = df_i["bbox_w"].to_numpy()
aspect = bbox_h / np.maximum(bbox_w, 1e-9)
angle = df_i["angle_deg"].to_numpy()

kin = pd.DataFrame({
    "t": df_i["t"],
    "speed_px_s": speed,
    "acc_px_s2": acc,
    "jerk_px_s3": jerk,
    "disp_px": disp,
    "path_len_px": path_len,
    "bbox_h_px": bbox_h,
    "bbox_w_px": bbox_w,
    "aspect_ratio": aspect,
    "angle_deg": angle
})

# 워밍업 1초 제외
warmup = int(round(fps * 1.0))
kin2 = kin.iloc[warmup:].reset_index(drop=True)

# 활동 프레임 정의
active = kin2[kin2["speed_px_s"] > 5].copy()

def summarize(s):
    s = pd.Series(s).dropna()
    return pd.Series({
        "N": int(s.shape[0]),
        "Mean": float(s.mean()),
        "Std": float(s.std(ddof=1)),
        "Min": float(s.min()),
        "Median": float(s.median()),
        "Max": float(s.max()),
        "P95": float(s.quantile(0.95)),
    })

# 전체 요약
vars_to_report = {
    "Speed (px/s)": kin2["speed_px_s"],
    "Acceleration (px/s^2)": kin2["acc_px_s2"],
    "Jerk (px/s^3)": kin2["jerk_px_s3"],
    "Displacement from start (px)": kin2["disp_px"],
    "Path length (px)": kin2["path_len_px"],
    "BBox height (px)": kin2["bbox_h_px"],
    "BBox width (px)": kin2["bbox_w_px"],
    "Aspect ratio (h/w)": kin2["aspect_ratio"],
    "Body axis angle (deg)": kin2["angle_deg"],
}

summary_tbl = pd.DataFrame({k: summarize(v) for k, v in vars_to_report.items()}).T
print("\n--- Summary (all frames, warmup removed) ---")
print(summary_tbl.round(2).to_markdown())

# 활동 프레임 요약
summary_active = pd.DataFrame({
    "Speed (px/s)": summarize(active["speed_px_s"]),
    "Acceleration (px/s^2)": summarize(active["acc_px_s2"]),
    "Jerk (px/s^3)": summarize(active["jerk_px_s3"]),
    "BBox height (px)": summarize(active["bbox_h_px"]),
    "Body axis angle (deg)": summarize(active["angle_deg"]),
}).T

print("\n--- Summary (active frames: speed > 5 px/s) ---")
print(summary_active.round(2).to_markdown())

# ---------
# 4) 그래프
# ---------
t = kin2["t"].to_numpy()

plt.figure(figsize=(6, 3))
plt.plot(t, kin2["speed_px_s"].to_numpy())
plt.xlabel("Time (s)")
plt.ylabel("Speed (px/s)")
plt.title("Center-of-mass proxy speed over time")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.plot(t, kin2["bbox_h_px"].to_numpy())
plt.xlabel("Time (s)")
plt.ylabel("Bounding box height (px)")
plt.title("Vertical extent proxy over time")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.plot(t, kin2["angle_deg"].to_numpy())
plt.xlabel("Time (s)")
plt.ylabel("Body axis angle (deg)")
plt.title("Body orientation proxy over time")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.hist(active["speed_px_s"].to_numpy(), bins=15)
plt.xlabel("Speed (px/s)")
plt.ylabel("Count")
plt.title("Speed distribution (active frames)")
plt.tight_layout()
plt.show()
```
